# **875**

```python
class Solution:
    def minEatingSpeed(self, piles: List[int], h: int) -> int:

        right = max(piles)
        left = 1

        while left < right:
            mid = (left + right) // 2

            cnt = 0
            for pile in piles:
                cnt += (pile-1) // mid + 1

            if cnt > h:
                left = mid + 1
            else:
                right = mid


        return left
```

# **410**

```python
class Solution:
    def splitArray(self, nums: List[int], k: int) -> int:
        l, r = max(nums), sum(nums)
        ans = 0

        while l <= r:
            mid = (l+r) // 2

            s, cnt = 0, 1
            for num in nums:
                if s + num > mid:
                    cnt += 1
                    s = num
                else:
                    s += num

            if cnt > k:
                l = mid + 1
            else:
                ans = mid
                r = mid - 1

        return ans
```
