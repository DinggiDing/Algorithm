```python
class Solution:
    def minEatingSpeed(self, piles: List[int], h: int) -> int:

        right = max(piles)
        left = 1

        while left < right:
            mid = (left + right) // 2

            cnt = 0
            for pile in piles:
                cnt += (pile-1) // mid + 1

            if cnt > h:
                left = mid + 1
            else:
                right = mid


        return left
```
