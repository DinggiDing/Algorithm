# **의상**

```python
import Foundation

func solution(_ clothes:[[String]]) -> Int {
    var n = clothes.count

    var dict = [String: Int]()

    for i in 0..<n {
        if dict.contains(where: { $0.key == clothes[i][1]}) {
            dict[clothes[i][1]]! += 1
        } else {
            dict[clothes[i][1]] = 1
        }
    }

    var answer : Int = 1

    for i in dict.keys {
        answer *= (dict[i]! + 1)
    }

    return answer-1
}
```

# **베스트앨범**

```python
import Foundation

func solution(_ genres:[String], _ plays:[Int]) -> [Int] {

    let n = genres.count

    var answer: [Int] = []
    var dict: [String: [[Int]]] = [:]

    for idx in 0..<n {
        if let _ = dict[genres[idx]] {
            dict[genres[idx]]?.append([idx, plays[idx]])
        } else {
            dict[genres[idx]] = [[idx, plays[idx]]]
        }
    }

    let sortedDict = dict.sorted { (first, second) -> Bool in
        return first.value.map { $0[1] }.reduce(0, +) > second.value.map { $0[1] }.reduce(0, +)
    }

    print(sortedDict)

    for idx in 0..<sortedDict.count {
        let item = sortedDict[idx].value.sorted { (first, second) -> Bool in
            return first[1] == second[1] ? false : first[1] > second[1]
        }

        for idx in 0..<item.count {
            if idx > 1 {
                break
            }

            answer.append(item[idx][0])
        }
    }

    return answer
}
```

# **타겟 넘버**

```python
import Foundation

func solution(_ numbers:[Int], _ target:Int) -> Int {
    var answer = 0
    let sum = numbers.reduce(0, +)
    let n = numbers.count

    if sum==target {
        return 1
    }

    func dfs(_ index: Int, _ brr: [Int]) {
        let p = brr.reduce(0, +)
        if (sum - 2*p) == target {
            answer += 1
        }


        for i in index..<numbers.count {
            dfs(i + 1, brr + [numbers[i]])
        }
    }

    dfs(0, [])

    return answer
}
```

# **네트워크**

```python
import Foundation

func solution(_ n:Int, _ computers:[[Int]]) -> Int {
    var answer = 0

    let n = computers.count

    var visited = Array(repeating: false, count: n)
    var arr = Array(repeating: 0, count: n)

    var queue = [Int]()

    for i in 0..<n {
        if visited[i] == false {
            answer += 1
            queue.append(i)
            // visited[0] = true

            while !queue.isEmpty {
                let new = queue.removeFirst()

                for i in 0..<n {
                    if visited[i] == false && computers[new][i] == 1 {
                        queue.append(i)
                        visited[i] = true
                        arr[i] = answer
                    }
                }
            }
        }
    }

    print(arr)

    return arr.max()!
}
```

# **디스크 컨트롤러**

```python
import Foundation

func solution(_ jobs:[[Int]]) -> Int {

    var sortedJobs = jobs.sorted(by: { $0[1] == $1[1] ? $0[0] < $1[0] : $0[1] < $1[1] })
    var now = 0
    var totalTime = 0

    while !sortedJobs.isEmpty {
        var check = false
        for i in 0..<sortedJobs.count {
            if sortedJobs[i][0] <= now {
                now += sortedJobs[i][1]
                totalTime += now - sortedJobs[i][0]
                sortedJobs.remove(at: i)
                check = true
                break
            }
        }
        if !check {
            now += 1
        }
    }

    return totalTime / jobs.count
}
```

# **섬 연결하기**

크루스칼 알고리즘

```python
import Foundation

func solution(_ n:Int, _ costs:[[Int]]) -> Int {

    var visited = Array(repeating: -1, count: n)
    let arr = costs.sorted(by: { $0[2] < $1[2] })
    var result = 0

    for i in 0..<n {
        visited[i] = i
    }

    for cost in arr {
        if find(&visited, cost[0]) != find(&visited, cost[1]) {
            union(&visited, cost[0], cost[1])
            result += cost[2]
        }
    }

    return result
}

func find(_ arr: inout [Int], _ x: Int) -> Int {
    if arr[x] != x {
        arr[x] = find(&arr, arr[x])
    }

    return arr[x]
}

func union(_ arr: inout [Int], _ a: Int, _ b: Int) {
    let a = find(&arr, a)
    let b = find(&arr, b)

    if a < b {
        arr[b] = a
    } else {
        arr[a] = b
    }
}
```

# **N으로 표현**

```python
import Foundation

func solution(_ N:Int, _ number:Int) -> Int {

    var arr:[[Int]] = Array(repeating: [], count: 9)

    for i in 1..<9 {
        var set = Set<Int>()
        set.insert(Int(String(repeating: String(N), count: i))!)

        for j in 1..<i {
            for cons1 in arr[i-j] {
                for cons2 in arr[j] {
                    let plus = cons1 + cons2
                    let minus = cons1 - cons2
                    let mul = cons1 * cons2
                    if cons2 != 0 && cons1 != 0 && cons1 >= cons2 {
                        let div = cons1 / cons2
                        set.insert(div)
                    }
                    if minus >= 0 {
                        set.insert(minus)
                    }
                    set.insert(plus)
                    set.insert(mul)
                }
            }
        }

        if set.contains(number) {
            return i
        }

        for s in set {
            arr[i].append(s)
        }
    }

    return -1
}
```

# **입국 심사**

```python
import Foundation

func solution(_ n:Int, _ times:[Int]) -> Int64 {
    let times = times.sorted(by: { $0 < $1 })
    var right = times.last! * n
    var left = 1


    while left <= right {
        var mid = (left + right)/2

        var result = 0
        for i in times {
            result += mid / i
        }

        if result < n {
            left = mid + 1
        } else {
            right = mid - 1
        }
    }


    return Int64(left)
}
```

# **문자열 압축**

```python
import Foundation

func solution(_ s:String) -> Int {
    let arr = Array(s)
    let nn = s.count
    let n = s.count / 2
    // comp += arr[0..<n]
    var result = [Int]()

    for i in 1...n {
        var comp = ""
        var count = 1
        var temp = arr[0..<i]

        for j in stride(from: i, to: nn, by: i) {
            var idx = j+i
            if idx > nn-1 {
                idx = arr.endIndex
            }

            if temp == arr[j..<idx] {
                count += 1
            } else {
                if count != 1 {
                    comp += "\(count)" + temp
                } else {
                    comp += temp
                }
                count = 1
                temp = arr[j..<idx]
            }
        }
        if count != 1 {
            comp += "\(count)" + temp
        } else {
            comp += temp
        }
        // print(i, comp)
        result.append(comp.count)
    }

    return result.min()!
}
```
