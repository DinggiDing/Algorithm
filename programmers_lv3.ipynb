```python
def solution(s):
    answer = 0

    n = len(s)
    for_off = False
    i_n = n

    while True:
      if for_off:
        break
      na = 1
      if n%2==0:
        na = 0
      for i in range(0, i_n - n + 1):
        arr = s[i:n//2+i]
        if arr[::-1]==s[n//2+na+i:(n//2)*2+na+i]:
          answer = n
          for_off = True
          break

      n -= 1

    return answer


s = "abadaba"
solution(s)
```

# **선입 선출 스케일링** : 다시 풀기

```python
def solution(n, cores):
    answer = 0

    number = len(cores)
    queue = [1 for i in range(number)]
    for_on = True

    while True:
      if n==0:
        break

      if for_on:
        for j in range(number):
          if queue[j]==1:
            queue[j] = cores[j]
            n -= 1
            if n==0:
              for_on = False
              answer = j+1
              break
          else:
            queue[j] -= 1

      else:
        break

    return answer


n = 6
cores = [1,2,3]
solution(n, cores)
```

# **야근 지수** : 다시 풀기

```python
from heapq import heapify, heappush, heappop

def solution(n, works):
    answer = 0

    if sum(works) <= n:
        return 0

    works = [(-1)*i for i in works]
    heapify(works)

    while n:
      p = heappop(works)
      q = p + 1
      heappush(works, q)
      n -= 1

    answer = sum([i*i for i in works])
    return answer


works = [10,10,1]
n = 10
solution(n, works)
```

# **최고의 집합**

```python
def solution(n, s):
    answer = []

    p = s // n
    q = s % n

    if n>s:
      answer.append(-1)
    else:
      for i in range(n):
        if i < (n-q):
          answer.append(p)
        else:
          answer.append(p+1)

    return answer


n = 3
s = 1
solution(n, s)
```

동적계획법

```python
def solution(matrix_sizes):
    n = len(matrix_sizes)

    # dp[i][j]: i부터 j까지의 행렬을 곱하기 위한 최소 곱셈 연산 수
    dp = [[0] * n for _ in range(n)]

    # 연쇄 행렬 곱셈의 길이에 따라 계산
    for chain_length in range(2, n + 1):
        for i in range(n - chain_length + 1):
            j = i + chain_length - 1
            dp[i][j] = float('inf')  # 무한대로 초기화
            print(i, j)

            for k in range(i, j):
                cost = dp[i][k] + dp[k+1][j] + matrix_sizes[i][0] * matrix_sizes[k][1] * matrix_sizes[j][1]
                dp[i][j] = min(dp[i][j], cost)

    return dp[0][n-1]

# Example
matrix_sizes = [[4, 7], [7, 100], [100, 4], [4, 1]]
result = solution(matrix_sizes)
print(result)
```

c : max(a+c, b)
d : max(a+d, a+c, b+d) = max(a+d, c, b+d) {a+c = c}
  = max(b+d, c) {max(a+d, b+d) = d + max(a, b) = b+d}

e : if a=check => d, a!=check => max(c+e, d)

> 규칙 발견
  index i 일때, c(index 2)이상부터는 max(i-2+i, i-1) 이다.
  i가 마지막이라면 a(index 0)인 요소가 sum에 포함되어있다면 i-1, 아니라면 max(i-2+i, i-1).


```python
def solution(sticker):
    answer = 0

    n = len(sticker)
    sticker = [sticker.copy() for i in range(n)]

    for p in range(n):

      dp = [False for i in range(n)]

      if sticker[p][(p+1)%n]>=sticker[p][p]:
        sticker[p][(p+1)%n] = sticker[p][(p+1)%n]
      else:
        sticker[p][(p+1)%n] = sticker[p][p]
        dp[1] = True
        dp[0] = True

      for i in range(2, n):
        if i==n-1:
          if dp[i-1]==True:
            sticker[p][(i+p)%n] = sticker[p][(i-1+p)%n]
          else:
            sticker[p][(i+p)%n] = max(sticker[p][(i-2+p)%n]+sticker[p][(i+p)%n], sticker[p][(i-1+p)%n])
        else:
          if i==2:
            if sticker[p][(i-2+p)%n]+sticker[p][(i+p)%n] >= sticker[p][(i-1+p)%n]:
              sticker[p][(i+p)%n] = sticker[p][(i-2+p)%n]+sticker[p][(i+p)%n]
              dp[i] = dp[i-2] = True
            else:
              sticker[p][(i+p)%n] = sticker[p][(i-1+p)%n]
              dp[i] = dp[i-1]
          else:
            if sticker[p][(i-2+p)%n]+sticker[p][(i+p)%n] >= sticker[p][(i-1+p)%n]:
              sticker[p][(i+p)%n] = sticker[p][(i-2+p)%n]+sticker[p][(i+p)%n]
              dp[i] = dp[i-2]
            else:
              sticker[p][(i+p)%n] = sticker[p][(i-1+p)%n]
              dp[i] = dp[i-1]

    #   print(p, sticker)
    # print(sticker)
    answer = max(sticker)
    return max(answer)


sticker = [20, 23]
solution(sticker)
```

```python
def solution(sticker):

    if len(sticker) == 1:
        return sticker[0]
    dp = [0 for _ in range(len(sticker))]  # use first sticker
    dp2 = [0 for _ in range(len(sticker))]  # unused first sticker

    dp[0] = sticker[0]
    dp[1] = sticker[0]

    dp2[0] = 0
    dp2[1] = sticker[1]

    for i in range(2, len(sticker) - 1):
        dp[i] = max(dp[i - 2] + sticker[i], dp[i - 1])

    value_1 = max(dp)

    for i in range(2, len(sticker)):
        dp2[i] = max(dp2[i - 2] + sticker[i], dp2[i - 1])
    value_2 = max(dp2)

    answer = max(value_1,value_2)
    return answer
```

```python
def solution(A, B):
    answer = 0

    A.sort(reverse=True)
    B.sort(reverse=True)

    n = len(A)
    a=0
    while True:
      if a==n:
        break
      if A[0] >= B[0]:
        del B[len(B)-1]
        del A[0]
      else:
        answer += 1
        del B[0]
        del A[0]
      a += 1
      # print(A,B)

    return answer


A, B = [2,2,2,2],	[1,1,1,1]
solution(A, B)
```

# **추석 트래픽**

```python
def solution(lines):
    answer = 0

    n = len(lines)
    q = [0 for i in range(n)]     # 시작시간
    p = [0 for i in range(n)]     # 종료시간

    for i in range(n):
      str_a = lines[i].split(" ")

      pp = str_a[1]
      hour = int(pp.split(":")[0])
      minute = int(pp.split(":")[1])
      second = int(pp.split(":")[2].split(".")[0])*1000 + int(pp.split(":")[2].split(".")[1])
      p[i] = (hour*60+minute)*60*1000+second
      q[i] = int(p[i] - float(str_a[2].split("s")[0])*1000) + 1
      p[i] += 1000


    for i in range(n):
      cnt = 0
      for j in range(i, n):
        if p[i] > q[j]:
          cnt += 1
      answer = max(answer, cnt)

    return answer


lines = [
"2016-09-15 20:59:57.421 0.351s",
"2016-09-15 20:59:58.233 1.181s",
"2016-09-15 20:59:58.299 0.8s",
"2016-09-15 20:59:58.688 1.041s",
"2016-09-15 20:59:59.591 1.412s",
"2016-09-15 21:00:00.464 1.466s",
"2016-09-15 21:00:00.741 1.581s",
"2016-09-15 21:00:00.748 2.31s",
"2016-09-15 21:00:00.966 0.381s",
"2016-09-15 21:00:02.066 2.62s"
]
solution(lines)
```

# **연속 펄스 부분 수열의 합**

```python
def solution(sequence):
    n = len(sequence)

    a = []
    b = []
    p = 1
    for i in sequence:
        a.append(i * p)
        b.append(i * -p)
        p *= -1

    aa = [0]
    bb = [0]
    for i in a:
        aa.append(max(0, aa[-1]) + i)

    for i in b:
        bb.append(max(0, bb[-1]) + i)
    return max(max(aa), max(bb))

```

# **보석 쇼핑**

투포인터

```python
def solution(gems):
    n = len(gems)
    st = set(gems)
    cnt = len(st)
    start, end = 0, 0

    dict = {gems[0]:1}
    answer = [0, n-1]

    while start <= end and end <= n:

        if len(dict) < cnt:
            end += 1
            if end==n:
                break
            dict[gems[end]] = dict.get(gems[end], 0) + 1
        else:
            if end - start < answer[1] - answer[0]:
                answer = [start, end]
            if dict[gems[start]]==1:
                del dict[gems[start]]
            else:
                dict[gems[start]] -= 1
            start += 1

        # answer = [s, e]
    answer[0] += 1
    answer[1] += 1
    return answer
```

# **경주로 건설**

DP + BFS

```python
def solution(board):

    n = len(board)
    MAX = float('inf')
    answer = MAX
    visited = [[MAX for _ in range(n)] for _ in range(n)]

    from collections import deque
    dx = [1, 0, -1, 0]
    dy = [0, 1, 0, -1]

    queue = deque()
    queue.append((0, 0, -1, 0))

    while queue:
        x, y, d, cost = queue.popleft()

        if x==n-1 and y==n-1 and answer > cost:
            answer = cost

        for i in range(4):
            nd = i
            nx = x + dx[nd]
            ny = y + dy[nd]

            nk = 1
            if d == -1:
                nk = 1
            else:
                nk = 1 if d==nd else 6

            if 0<= nx <n and 0<= ny <n and board[ny][nx] == 0:
                if visited[ny][nx] >= cost + nk - 4:
                    visited[ny][nx] = cost + nk
                    queue.append((nx, ny, nd, cost + nk))

    return answer*100
```

# **합승 택시 요금**

그래프 (플로이드)

```python
def solution(n, s, a, b, fares):
    MAX = float('inf')

    arr = [MAX for _ in range(n+1)]
    graph = [[MAX for _ in range(n+1)] for _ in range(n+1)]

    for fare in fares:
        graph[fare[0]][fare[1]] = graph[fare[1]][fare[0]] = fare[2]

    def floyd_warshall(graph):
        for k in range(1, n+1):
            for a in range(1, n+1):
                for b in range(1, n+1):
                    graph[a][b] = min(graph[a][b], graph[a][k] + graph[k][b])

    floyd_warshall(graph)
    for i in range(1, n+1):
        graph[i][i] = 0
    # for i in graph:
    #     print(i)

    for i in range(1, n+1):
        arr[i] = graph[s][i] + graph[i][b] + graph[i][a]

    # print("--- * ---")
    # print(arr)
    return min(arr)
```

# **다단계 칫솔 판매**

딕셔너리, 구현

```python
def solution(enroll, referral, seller, amount):

    dic = dict(zip(enroll, referral))
    diccnt = {}

    answer = []

    for i in range(len(seller)):
        amt = amount[i]*100
        sel = seller[i]

        while amt > 0:
            rest = amt
            amt = amt // 10
            rest -= amt
            diccnt[sel] = diccnt.get(sel, 0) + rest

            sel = dic[sel]

            if sel == "-":
                break

    # print(diccnt)
    for i in enroll:
        answer.append(diccnt.get(i, 0))

    return answer
```
