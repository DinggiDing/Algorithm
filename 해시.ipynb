# **완주하지 못한 선수**

```python
def solution(participant, completion):
    answer = ''

    hash = {}
    for i in participant:
      if i in hash:
        hash[i] += 1
      else:
        hash[i] = 1
    for i in completion:
      if hash[i]!=0:
        hash[i] -= 1
      if hash[i]==0:
        del hash[i]

    answer = list(hash.keys())[0]
    return answer

participant = ["marina", "josipa", "nikola", "vinko", "filipa"]
completion = ["josipa", "filipa", "marina", "nikola"]
solution(participant, completion)
```

# **의상**

5*1*1 => 5+1+1 + 5+5+1 + 5 = 23
2*1*1 => 2+1+1 + 2+2+1 + 2 = 11
2*1   => 2+1 + 2 = 5
3     => 3

```python
def solution(clothes):
    answer = 1

    hash = {}
    for i in clothes:
      if i[1] in hash:
        hash[i[1]] += 1
      else:
        hash[i[1]] = 1

    for i in list(hash.values()):
      answer *= i+1
    return answer-1


clothes = [["yellow_hat", "headgear"], ["blue_sunglasses", "eyewear"], ["green_turban", "headgear"]]
solution(clothes)
```

# **베스트앨범**

```python
def solution(genres, plays):
    answer = []

    hash = {}
    sum = {}

    for i in range(len(genres)):
      gen = genres[i]
      pla = plays[i]
      if gen in hash:
        hash[gen] += pla
        sum[gen].append([pla, i])
      else:
        hash[gen] = pla
        sum[gen] = [[pla, i]]

    sum_arr = list(hash.items())
    sum_arr.sort(key=lambda x: -x[1])

    arr= []
    for i in range(len(sum_arr)):
      arr.append(sum[sum_arr[i][0]])

    for i in range(len(arr)):
      arr[i].sort(key=lambda x: -x[0])

    for i in range(len(arr)):
      for j in range(len(arr[i])):
        if j<2:
          answer.append(arr[i][j][1])

    return answer

genres = ["classic", "pop", "classic", "classic", "pop"]
plays = [500, 600, 150, 800, 2500]
solution(genres, plays)
```

# **더 맵게**

```python
def solution(scoville, K):
    answer = 0

    n = 0

    import heapq

    heapq.heapify(scoville)

    while True:
      a = heapq.heappop(scoville)
      print(a)
      if a >= K:
        break
      if len(scoville) < 1:
        n = -1
        break

      b = heapq.heappop(scoville)
      c = a + 2*b
      heapq.heappush(scoville, c)
      n += 1


    return n


scoville = [2,3]
K = 7
solution(scoville, K)
```

# **이중우선순위큐**

```python
def solution(operations):
    answer = []

    import heapq

    heap = []

    heapq.heapify(heap)
    for item in operations:
      a, b = item.split()
      b = int(b)
      print(a,b)

      if a == "I":
        heapq.heappush(heap, b)
      else:
        if heap:
          if b == -1:
            heapq.heappop(heap)
          else:
            heap.sort(reverse=True)
            heapq.heappop(heap)

    heap.sort()
    if heap==[]:
      answer = [0, 0]
    else:
      answer = [heap[-1], heap[0]]

    return answer


operations = ["I -45", "I 653", "D 1", "I -642", "I 45", "I 97", "D 1", "D -1", "I 333"]
solution(operations)
```
