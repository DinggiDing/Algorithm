# **Two Sum**

Python

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        answer = []
        for i in range(len(nums)-1):
            for j in range(i+1, len(nums)):
                if nums[i] + nums[j] == target:
                    answer.append(i)
                    answer.append(j)
                    break

        return answer
```

# **Palindrome Number**

Python

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        a = str(x)
        n = len(a)//2 if len(a)%2==0 else len(a)//2+1
        answer = True

        for i in range(n):
            p = a[i]
            q = a[-(i+1)]
            if p!=q:
                answer = False

        return answer
```

# **Flood Fill**

```python
class Solution {
    func floodFill(_ image: [[Int]], _ sr: Int, _ sc: Int, _ color: Int) -> [[Int]] {

        let m = image[0].count
        let n = image.count
        var visited = Array(repeating: Array(repeating: false, count: m), count: n)
        var queue = [(Int, Int)]()
        var arr = image

        let start = (sr, sc)
        let point_color = image[sr][sc]
        queue.append(start)
        visited[sr][sc] = true
        arr[sr][sc] = color

        while !queue.isEmpty {
            let new = queue.removeFirst()

            let dx = [1, -1, 0, 0]
            let dy = [0, 0, -1, 1]

            for i in 0..<4 {
                let nx = new.0 + dx[i]
                let ny = new.1 + dy[i]

                if 0 <= ny && ny < m && 0 <= nx && nx < n && image[nx][ny] == point_color && visited[nx][ny]==false {

                    queue.append((nx, ny))
                    visited[nx][ny] = true
                    arr[nx][ny] = color
                }
            }
        }

        return arr
    }
}
```

# **Number of Wonderful Substring** : 다시풀기

```python

class Solution:
  def wonderfulSubstrings(self, word: str) -> int:
    ans = 0
    prefix = 0  # the binary prefix
    count = [0] * 1024  # the binary prefix count
    count[0] = 1  # the empty string ""

    for c in word:
      prefix ^= 1 << ord(c) - ord('a')
      # All the letters occur even number of times.
      ans += count[prefix]
      # `c` occurs odd number of times.
      ans += sum(count[prefix ^ 1 << i] for i in range(10))
      count[prefix] += 1

    return ans
```

# **Reverse Prefix of Word**

```python
class Solution:
    def reversePrefix(self, word: str, ch: str) -> str:

        stack = ""
        idx = -1

        for i in range(len(word)):
            if word[i] == ch:
                stack += word[i]
                idx = i+1
                break
            else:
                stack += word[i]

        if idx != -1:
            stackarr = list(stack)
            stackarr.reverse()
            stack = ''.join(stackarr)
            stack += word[idx:]
        else:
            stack = word

        return stack
```

```python
class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:

        answer = True

        # row
        for i in range(9):

            visited = [False for _ in range(10)]

            for j in range(9):
                if board[i][j] != ".":
                    if visited[int(board[i][j])]==False:
                        visited[int(board[i][j])] = True
                    else:
                        answer = False


        # column
        for i in range(9):

            visited = [False for _ in range(10)]

            for j in range(9):
                if board[j][i] != ".":
                    if visited[int(board[j][i])]==False:
                        visited[int(board[j][i])] = True
                    else:
                        answer = False

        # 3*3
        for i in range(9):

            visited = [False for _ in range(10)]

            for j in range(9):
                a = (i // 3)*3 + j // 3
                b = (i % 3)*3 + j % 3
                if board[a][b] != ".":
                    if visited[int(board[a][b])]==False:
                        visited[int(board[a][b])] = True
                    else:
                        answer = False


        return answer
```

# **165 Compare Version Numbers**

```python
class Solution:
    def compareVersion(self, version1: str, version2: str) -> int:

        answer = 0

        arr = version1.split('.')
        arr = list(map(int, arr))

        brr = version2.split('.')
        brr = list(map(int, brr))

        na = len(arr)
        nb = len(brr)

        if na < nb:
            for i in range(nb - na):
                arr.append(0)
        else:
            for i in range(na - nb):
                brr.append(0)

        for i in range(len(arr)):
            if arr[i] < brr[i]:
                answer = -1
                break
            if arr[i] > brr[i]:
                answer = 1
                break
            if arr[i] == brr[i] and i==len(arr)-1:
                answer = 0

        return answer
```

```python
class Solution:
    def spiralOrder(self, matrix: List[List[int]]) -> List[int]:

        dx = [1, 0, -1, 0]
        dy = [0, 1, 0, -1]

        m = len(matrix[0])
        n = len(matrix)
        visited = [[False for _ in range(m)] for _ in range(n)]
        answer = []
        i = 0
        direct = 0
        x = 0
        y = 0
        visited[0][0] = True
        answer.append(matrix[y][x])
        while i<m*n-1:

            if direct == 4:
                direct = 0

            nx = x + dx[direct]
            ny = y + dy[direct]

            if 0 <= nx < m and 0 <= ny < n and visited[ny][nx]==False:
                x = nx
                y = ny
                visited[ny][nx] = True
                answer.append(matrix[ny][nx])
            else:
                direct += 1
                if direct == 4:
                    direct = 0
                nx = x + dx[direct]
                ny = y + dy[direct]
                x = nx
                y = ny
                visited[ny][nx] = True
                answer.append(matrix[ny][nx])

            i += 1

        return answer
```

```python
class Solution:
    def numIslands(self, grid: List[List[str]]) -> int:

        m = len(grid[0])
        n = len(grid)
        visited = [[-1 for _ in range(m)] for _ in range(n)]
        cnt = 0

        from collections import deque

        for a in range(n):
            for b in range(m):
                if grid[a][b] == "1" and visited[a][b]==-1:
                    cnt += 1
                    queue = deque()
                    queue.append((b, a))
                    visited[a][b] = cnt

                    dx = [1, -1, 0, 0]
                    dy = [0, 0, 1, -1]

                    while queue:
                        x, y = queue.pop()

                        for i in range(4):
                            nx = x + dx[i]
                            ny = y + dy[i]

                            if 0<=nx<m and 0<=ny<n and grid[ny][nx]=="1" and visited[ny][nx]==-1:
                                queue.append((nx, ny))
                                visited[ny][nx] = cnt

        return cnt
```

```python
class Solution:
    def kthSmallestPrimeFraction(self, arr: List[int], k: int) -> List[int]:

        from itertools import combinations
        answer = []
        dict = {}

        for i in combinations(arr, 2):
            answer.append(i[0]/ i[1])
            dict[i[0]/i[1]] = [i[0], i[1]]

        answer.sort(key=lambda x: x)

        return dict[answer[k-1]]
```

# **861 Score After Flipping Matrix**

```python
class Solution:
    def matrixScore(self, grid: List[List[int]]) -> int:

        m = len(grid[0])
        n = len(grid)

        while True:
            # 1. All 1st column => 1
            # 2. half of 2nd ~ column => 1

            for i in range(n):
                if grid[i][0] == 0:
                    for j in range(m):
                        grid[i][j] ^= 1

            for i in range(1, m):
                cnt = 0
                for j in range(n):
                    if grid[j][i]==1:
                        cnt += 1

                if cnt < n/2:
                    for j in range(n):
                        grid[j][i] ^= 1

            break

        answer = 0
        for i in range(n):
            st = ""
            for j in range(m):
                st += str(grid[i][j])
            answer += int(st, 2)

        return answer
```

# **1219**

```python
class Solution:
    def getMaximumGold(self, grid: List[List[int]]) -> int:
        m = len(grid[0])
        n = len(grid)

        visited = [[False for _ in range(len(grid[0]))] for _ in range(len(grid))]
        dx = [1, -1, 0, 0]
        dy = [0, 0, 1, -1]

        arr = [0]
        def dfs(nx: int, ny: int, s: int):
            if arr[0] < s:
                arr[0] = s
            for i in range(4):
                x = nx + dx[i]
                y = ny + dy[i]

                if 0<= x <m and 0<= y <n and visited[y][x]==False and grid[y][x]!=0:
                    visited[y][x] = True
                    dfs(x, y, s+grid[y][x])
                    visited[y][x] = False

        for i in range(n):
            for j in range(m):
                if grid[i][j]>0:
                    visited[i][j] = True
                    dfs(j, i, grid[i][j])
                    visited[i][j] = False

        # dfs(1, 2, 0)

        return arr[0]
```
