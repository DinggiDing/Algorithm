# **Two Sum**

Python

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        answer = []
        for i in range(len(nums)-1):
            for j in range(i+1, len(nums)):
                if nums[i] + nums[j] == target:
                    answer.append(i)
                    answer.append(j)
                    break

        return answer
```

Swift

```python
class Solution {
    func twoSum(_ nums: [Int], _ target: Int) -> [Int] {
        var answer = [Int]()
        for i in 0..<nums.count-1 {
            for j in i+1..<nums.count {
                if nums[i] + nums[j] == target {
                    answer.append(i)
                    answer.append(j)
                    break
                }
            }
        }

        return answer
    }
}
```

# **Palindrome Number**

Python

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        a = str(x)
        n = len(a)//2 if len(a)%2==0 else len(a)//2+1
        answer = True

        for i in range(n):
            p = a[i]
            q = a[-(i+1)]
            if p!=q:
                answer = False

        return answer
```

Swift

```python
class Solution {
    func isPalindrome(_ x: Int) -> Bool {
        let a = String(x)
        var answer = true

        var n = a.count%2==0 ? Int(a.count/2) : Int(a.count)/2 + 1

        for i in 0..<n {
            if a[a.index(a.startIndex, offsetBy: i)] != a[a.index(a.startIndex, offsetBy: a.count-i-1)] {
                answer = false
            }
        }
        return answer
    }
}
```

# **Roman to Integer**

```python
class Solution:
    def romanToInt(self, s: str) -> int:
        roman = {'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000,'IV':4,'IX':9,'XL':40,'XC':90,'CD':400,'CM':900}
        i = 0
        num = 0
        while i < len(s):
            if i+1<len(s) and s[i:i+2] in roman:
                num+=roman[s[i:i+2]]
                i+=2
            else:
                #print(i)
                num+=roman[s[i]]
                i+=1
        return num
```

# **Longest Common Prefix**

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        answer = ""

        for i in range(len(strs[0]), 0, -1):
            isit = 0
            for j in range(len(strs)-1):
                if strs[0][0:i] in strs[j+1][0:i]:
                    isit += 1
            if isit == len(strs)-1:
                answer = strs[0][0:i]
                break
        return answer
```

```python
class Solution {
    func longestCommonPrefix(_ strs: [String]) -> String {

        let n = strs.count
        var answer : String = ""
        var sum = 0
        for i in 0..<strs[0].count {
            sum = 0
            for j in 1..<strs.count {
                if String(strs[j].prefix(strs[0].count - i)) == String(strs[0].prefix(strs[0].count - i)) {
                    sum += 1
                }
            }
            if sum == strs.count-1 {
                answer = String(strs[0].prefix(strs[0].count - i))
                break
            }
        }

        return answer
    }
}
```

# **Island Perimeter**

```python
class Solution:
    def islandPerimeter(self, grid: List[List[int]]) -> int:

        from collections import deque

        answer = 0

        m = len(grid[0])
        n = len(grid)
        arr = [[-1 for _ in range(m)] for _ in range(n)]
        visited = [[False for _ in range(m)] for _ in range(n)]

        start = (0,0)
        is_for = False
        for i in range(n):
            if is_for:
                break
            for j in range(m):
                if grid[i][j]==1:
                    start = (j, i)
                    is_for = True
                    break

        print(start)
        queue = deque([])
        queue.append(start)
        visited[start[1]][start[0]] = True

        while queue:
            x, y = queue.popleft()

            dx = [1, -1, 0, 0]
            dy = [0, 0, 1, -1]

            cnt = 0
            for i in range(4):
                nx = x + dx[i]
                ny = y + dy[i]

                if 0<= nx <m and 0<= ny <n and grid[ny][nx]==1:
                    if visited[ny][nx]==False:
                        queue.append((nx, ny))
                        visited[ny][nx] = True
                else:
                    cnt += 1

            if cnt>0:
                arr[y][x] = cnt

        print(arr)
        for i in range(n):
            for j in range(m):
                if arr[i][j]>0:
                    answer += arr[i][j]

        return answer
```

# **Flood Fill**

```python
class Solution {
    func floodFill(_ image: [[Int]], _ sr: Int, _ sc: Int, _ color: Int) -> [[Int]] {

        let m = image[0].count
        let n = image.count
        var visited = Array(repeating: Array(repeating: false, count: m), count: n)
        var queue = [(Int, Int)]()
        var arr = image

        let start = (sr, sc)
        let point_color = image[sr][sc]
        queue.append(start)
        visited[sr][sc] = true
        arr[sr][sc] = color

        while !queue.isEmpty {
            let new = queue.removeFirst()

            let dx = [1, -1, 0, 0]
            let dy = [0, 0, -1, 1]

            for i in 0..<4 {
                let nx = new.0 + dx[i]
                let ny = new.1 + dy[i]

                if 0 <= ny && ny < m && 0 <= nx && nx < n && image[nx][ny] == point_color && visited[nx][ny]==false {

                    queue.append((nx, ny))
                    visited[nx][ny] = true
                    arr[nx][ny] = color
                }
            }
        }

        return arr
    }
}
```
