# **Two Sum**

Python

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        answer = []
        for i in range(len(nums)-1):
            for j in range(i+1, len(nums)):
                if nums[i] + nums[j] == target:
                    answer.append(i)
                    answer.append(j)
                    break

        return answer
```

Swift

```python
class Solution {
    func twoSum(_ nums: [Int], _ target: Int) -> [Int] {
        var answer = [Int]()
        for i in 0..<nums.count-1 {
            for j in i+1..<nums.count {
                if nums[i] + nums[j] == target {
                    answer.append(i)
                    answer.append(j)
                    break
                }
            }
        }

        return answer
    }
}
```

# **Palindrome Number**

Python

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        a = str(x)
        n = len(a)//2 if len(a)%2==0 else len(a)//2+1
        answer = True

        for i in range(n):
            p = a[i]
            q = a[-(i+1)]
            if p!=q:
                answer = False

        return answer
```

Swift

```python
class Solution {
    func isPalindrome(_ x: Int) -> Bool {
        let a = String(x)
        var answer = true

        var n = a.count%2==0 ? Int(a.count/2) : Int(a.count)/2 + 1

        for i in 0..<n {
            if a[a.index(a.startIndex, offsetBy: i)] != a[a.index(a.startIndex, offsetBy: a.count-i-1)] {
                answer = false
            }
        }
        return answer
    }
}
```

# **Roman to Integer**

```python
class Solution:
    def romanToInt(self, s: str) -> int:
        roman = {'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000,'IV':4,'IX':9,'XL':40,'XC':90,'CD':400,'CM':900}
        i = 0
        num = 0
        while i < len(s):
            if i+1<len(s) and s[i:i+2] in roman:
                num+=roman[s[i:i+2]]
                i+=2
            else:
                #print(i)
                num+=roman[s[i]]
                i+=1
        return num
```

```python
class Solution {
    func romanToInt(_ s: String) -> Int {
        let roman: [String: Int] = ["I":1,"V":5,"X":10,"L":50,"C":100,"D":500,"M":1000,"IV":4,"IX":9,"XL":40,"XC":90,"CD":400,"CM":900]

        var i = 0
        var answer = 0
        let str = Array(s)

        while i<s.count {

            if i+1 < s.count && (roman[String(str[i]) + String(str[i+1])] != nil) {
                answer += roman[String(str[i]) + String(str[i+1])]!
                i += 2

            } else {
                answer += roman[String(str[i])]!
                i += 1
            }
        }

        return answer
    }
}
```
