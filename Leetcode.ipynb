# **Two Sum**

Python

```python
class Solution:
    def twoSum(self, nums: List[int], target: int) -> List[int]:
        answer = []
        for i in range(len(nums)-1):
            for j in range(i+1, len(nums)):
                if nums[i] + nums[j] == target:
                    answer.append(i)
                    answer.append(j)
                    break

        return answer
```

# **Palindrome Number**

Python

```python
class Solution:
    def isPalindrome(self, x: int) -> bool:
        a = str(x)
        n = len(a)//2 if len(a)%2==0 else len(a)//2+1
        answer = True

        for i in range(n):
            p = a[i]
            q = a[-(i+1)]
            if p!=q:
                answer = False

        return answer
```

Swift

```python
class Solution {
    func isPalindrome(_ x: Int) -> Bool {
        let a = String(x)
        var answer = true

        var n = a.count%2==0 ? Int(a.count/2) : Int(a.count)/2 + 1

        for i in 0..<n {
            if a[a.index(a.startIndex, offsetBy: i)] != a[a.index(a.startIndex, offsetBy: a.count-i-1)] {
                answer = false
            }
        }
        return answer
    }
}
```

# **Roman to Integer**

```python
class Solution:
    def romanToInt(self, s: str) -> int:
        roman = {'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000,'IV':4,'IX':9,'XL':40,'XC':90,'CD':400,'CM':900}
        i = 0
        num = 0
        while i < len(s):
            if i+1<len(s) and s[i:i+2] in roman:
                num+=roman[s[i:i+2]]
                i+=2
            else:
                #print(i)
                num+=roman[s[i]]
                i+=1
        return num
```

# **Longest Common Prefix**

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        answer = ""

        for i in range(len(strs[0]), 0, -1):
            isit = 0
            for j in range(len(strs)-1):
                if strs[0][0:i] in strs[j+1][0:i]:
                    isit += 1
            if isit == len(strs)-1:
                answer = strs[0][0:i]
                break
        return answer
```

```python
class Solution {
    func longestCommonPrefix(_ strs: [String]) -> String {

        let n = strs.count
        var answer : String = ""
        var sum = 0
        for i in 0..<strs[0].count {
            sum = 0
            for j in 1..<strs.count {
                if String(strs[j].prefix(strs[0].count - i)) == String(strs[0].prefix(strs[0].count - i)) {
                    sum += 1
                }
            }
            if sum == strs.count-1 {
                answer = String(strs[0].prefix(strs[0].count - i))
                break
            }
        }

        return answer
    }
}
```

# **Island Perimeter**

```python
class Solution:
    def islandPerimeter(self, grid: List[List[int]]) -> int:

        from collections import deque

        answer = 0

        m = len(grid[0])
        n = len(grid)
        arr = [[-1 for _ in range(m)] for _ in range(n)]
        visited = [[False for _ in range(m)] for _ in range(n)]

        start = (0,0)
        is_for = False
        for i in range(n):
            if is_for:
                break
            for j in range(m):
                if grid[i][j]==1:
                    start = (j, i)
                    is_for = True
                    break

        print(start)
        queue = deque([])
        queue.append(start)
        visited[start[1]][start[0]] = True

        while queue:
            x, y = queue.popleft()

            dx = [1, -1, 0, 0]
            dy = [0, 0, 1, -1]

            cnt = 0
            for i in range(4):
                nx = x + dx[i]
                ny = y + dy[i]

                if 0<= nx <m and 0<= ny <n and grid[ny][nx]==1:
                    if visited[ny][nx]==False:
                        queue.append((nx, ny))
                        visited[ny][nx] = True
                else:
                    cnt += 1

            if cnt>0:
                arr[y][x] = cnt

        print(arr)
        for i in range(n):
            for j in range(m):
                if arr[i][j]>0:
                    answer += arr[i][j]

        return answer
```

# **Flood Fill**

```python
class Solution {
    func floodFill(_ image: [[Int]], _ sr: Int, _ sc: Int, _ color: Int) -> [[Int]] {

        let m = image[0].count
        let n = image.count
        var visited = Array(repeating: Array(repeating: false, count: m), count: n)
        var queue = [(Int, Int)]()
        var arr = image

        let start = (sr, sc)
        let point_color = image[sr][sc]
        queue.append(start)
        visited[sr][sc] = true
        arr[sr][sc] = color

        while !queue.isEmpty {
            let new = queue.removeFirst()

            let dx = [1, -1, 0, 0]
            let dy = [0, 0, -1, 1]

            for i in 0..<4 {
                let nx = new.0 + dx[i]
                let ny = new.1 + dy[i]

                if 0 <= ny && ny < m && 0 <= nx && nx < n && image[nx][ny] == point_color && visited[nx][ny]==false {

                    queue.append((nx, ny))
                    visited[nx][ny] = true
                    arr[nx][ny] = color
                }
            }
        }

        return arr
    }
}
```

# **Climbing Stairs**

DP 문제

```python
class Solution:
    def climbStairs(self, n: int) -> int:
        k = 0
        a = 1
        b = 2

        if n==1:
            b=a
            return b
        for i in range(n-2):
            k = b
            b += a
            a = k

        return b
```

# **Letter Combinations of a Phone Number**

```python
class Solution:
    def letterCombinations(self, digits: str) -> List[str]:
        arr = ["abc", "def", "ghi", "jkl", "mno", "pqrs", "tuv", "wxyz"]
        answer = []


        def dfs(row: int, output: str):
            if row==len(digits):
                answer.append(output)
                return

            for i in range(len(arr[int(digits[row])-2])):
                dfs(row+1, output + arr[int(digits[row])-2][i])

        if len(digits)!=0:
            dfs(0, "")

        return answer
```

# **Is Subsequence**

```python
class Solution {
    func isSubsequence(_ s: String, _ t: String) -> Bool {
        let n1 = t.count
        let n2 = s.count

        var cnt = 0
        var answer = false

        if n2 == 0 {
            return true
        }

        for i in 0..<n1 {

            if t[t.index(t.startIndex, offsetBy: i)] == s[s.index(s.startIndex, offsetBy: cnt)] {
                cnt += 1
            }

            if cnt == n2 {
                answer = true
                break
            }
        }



        return answer
    }
}
```

# **Search a 2D Matrix**

```python
class Solution {
    func searchMatrix(_ matrix: [[Int]], _ target: Int) -> Bool {
        let m = matrix[0].count
        let n = matrix.count

        var row = 0
        var answer = false
        for i in 0..<n {
            if matrix[i][0] <= target {
                row = i
            } else {
                break
            }
        }

        for i in 0..<m {
            if matrix[row][i] == target {
                answer = true
            }
        }

        return answer
    }
}
```

# **Find Peak Element**

```python
class Solution {
    func findPeakElement(_ nums: [Int]) -> Int {

        // return nums.firstIndex(of: nums.max()!)!

        var left = 0
        var right = nums.count - 1

        while left < right {
            let mid = (left + right) / 2

            if nums[mid] < nums[mid + 1] {
                left = mid + 1
            } else {
                right  = mid
            }
        }

        return left
    }
}
```

# **2997 Minimum Number of Operations to Make Array XOR equals to K**

```python
class Solution:
    def minOperations(self, nums: List[int], k: int) -> int:

        answer = 0
        a = 0
        for i in nums:
            a ^= i

        a ^= k

        b = str(format(a, 'b'))     # 2진수 변환
        for i in range(len(b)):
            if b[i]=="1":
                answer += 1

        return answer
```

# **Number of Wonderful Substring** : 다시풀기

```python

class Solution:
  def wonderfulSubstrings(self, word: str) -> int:
    ans = 0
    prefix = 0  # the binary prefix
    count = [0] * 1024  # the binary prefix count
    count[0] = 1  # the empty string ""

    for c in word:
      prefix ^= 1 << ord(c) - ord('a')
      # All the letters occur even number of times.
      ans += count[prefix]
      # `c` occurs odd number of times.
      ans += sum(count[prefix ^ 1 << i] for i in range(10))
      count[prefix] += 1

    return ans
```

# **Reverse Prefix of Word**

```python
class Solution:
    def reversePrefix(self, word: str, ch: str) -> str:

        stack = ""
        idx = -1

        for i in range(len(word)):
            if word[i] == ch:
                stack += word[i]
                idx = i+1
                break
            else:
                stack += word[i]

        if idx != -1:
            stackarr = list(stack)
            stackarr.reverse()
            stack = ''.join(stackarr)
            stack += word[idx:]
        else:
            stack = word

        return stack
```

# **Valid Sudoku**

```python
class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:

        answer = True

        # row
        for i in range(9):

            visited = [False for _ in range(10)]

            for j in range(9):
                if board[i][j] != ".":
                    if visited[int(board[i][j])]==False:
                        visited[int(board[i][j])] = True
                    else:
                        answer = False


        # column
        for i in range(9):

            visited = [False for _ in range(10)]

            for j in range(9):
                if board[j][i] != ".":
                    if visited[int(board[j][i])]==False:
                        visited[int(board[j][i])] = True
                    else:
                        answer = False

        # 3*3
        for i in range(9):

            visited = [False for _ in range(10)]

            for j in range(9):
                a = (i // 3)*3 + j // 3
                b = (i % 3)*3 + j % 3
                if board[a][b] != ".":
                    if visited[int(board[a][b])]==False:
                        visited[int(board[a][b])] = True
                    else:
                        answer = False


        return answer
```

# **2441 Largest Positive Integer That**

```python
class Solution:
    def findMaxK(self, nums: List[int]) -> int:

        answer = -1

        while True:
            if len(nums) == 0:
                break

            # print(nums.count)
            a = max(nums)
            if -a in nums:
                answer = a
                break
            else:
                del nums[nums.index(a)]

        return answer
```

# **1895 Largest Magic Square**

```python
class Solution:
    def largestMagicSquare(self, grid: List[List[int]]) -> int:

        answer = 1
        def checksum(arr: List[List[int]]) -> bool:
            n = len(arr)

            p_set = set()
            for i in arr:
                p_set.add(sum(i))
            if len(p_set) != 1:
                return False

            q_set = set()
            for i in zip(*arr):
                q_set.add(sum(i))
            if len(q_set) != 1:
                return False

            if p_set != q_set:
                return False

            r_a = 0
            r_b = 0
            for i in range(n):
                r_a += arr[i][i]
                r_b += arr[i][n-1-i]
            if r_a != r_b:
                return False
            if r_a != list(p_set)[0]:
                return False

            print(p_set, q_set, r_a)

            return True

        cnt = min(len(grid), len(grid[0]))
        for k in range(cnt, 0, -1):
            for i in range(len(grid) - k + 1):
                for j in range(len(grid[0]) - k + 1):
                    if checksum([row[j:j+k] for row in grid[i:i+k]])==True:
                        answer = k
                        return answer

        return answer
```
