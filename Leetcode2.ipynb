# **238**

누적합

```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:

        ans = [1]*len(nums)
        wer = [1]*len(nums)
        answer = [0] * len(nums)

        # From LEFT'
        ans[0] = nums[0]
        for i in range(1, len(nums)):
            ans[i] = ans[i-1] * nums[i]

        # From RIGHT
        wer[-1] = nums[-1]
        for i in range(len(nums)-2, -1, -1):
            wer[i] = wer[i+1] * nums[i]

        answer[0] = wer[1]
        answer[-1] = ans[-2]
        for i in range(1, len(nums)-1):
            answer[i] = ans[i-1] * wer[i+1]

        return answer
```

# **1002**

구현

```python
class Solution:
    def commonChars(self, words: List[str]) -> List[str]:

        ans = []
        win = list(words[0])

        for word in words:
            for i in range(len(word)):
                if word[i] in win:
                    ans.append(word[i])
                    # print(win[win.index(word[i])])
                    del win[win.index(word[i])]

            win = ans
            ans = []

        return win
```

# **1657**

딕셔너리

```python
class Solution:
    def closeStrings(self, word1: str, word2: str) -> bool:

        w1 = {}
        w2 = {}

        for i in range(len(word1)):
            w1[word1[i]] = w1.get(word1[i], 0) + 1

        for i in range(len(word2)):
            w2[word2[i]] = w2.get(word2[i], 0) + 1

        a1 = sorted(w1.values())
        a2 = sorted(w2.values())
        b1 = sorted(w1.keys())
        b2 = sorted(w2.keys())

        # print(a1, b1)
        # print(a2, b2)

        if a1 == a2 and b1 == b2:
            return True
        else:
            return False
```

# **648**

구현

```python
class Solution:
    def replaceWords(self, dic: List[str], sentence: str) -> str:

        sent = sentence.split(' ')

        di = sorted(dic, key=lambda x: len(x), reverse=False)

        for i in range(len(sent)):
            for j in di:

                if j == sent[i][0:len(j)]:
                    sent[i] = j
                    break

        return ' '.join(sent)
```

# **2390**

큐

```python
class Solution:
    def removeStars(self, s: str) -> str:
        from collections import deque

        queue = deque()

        for i in range(len(s)):
            if s[i] == "*":
                queue.pop()
            else:
                queue.append(s[i])

        # print(queue)

        return ''.join(queue)
```
