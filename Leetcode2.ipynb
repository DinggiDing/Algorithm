```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:

        ans = [1]*len(nums)
        wer = [1]*len(nums)
        answer = [0] * len(nums)

        # From LEFT'
        ans[0] = nums[0]
        for i in range(1, len(nums)):
            ans[i] = ans[i-1] * nums[i]

        # From RIGHT
        wer[-1] = nums[-1]
        for i in range(len(nums)-2, -1, -1):
            wer[i] = wer[i+1] * nums[i]

        answer[0] = wer[1]
        answer[-1] = ans[-2]
        for i in range(1, len(nums)-1):
            answer[i] = ans[i-1] * wer[i+1]

        return answer
```

```python
class Solution:
    def checkSubarraySum(self, nums: List[int], k: int) -> bool:

        # prefix sum
        seen, cur = {0: -1}, 0
        for i, a in enumerate(nums):
            cur = (cur + a) % k if k else cur + a
            if i - seen.setdefault(cur, i) > 1: return True
        return False
```

```python
class Solution:
    def subarraysDivByK(self, nums: List[int], k: int) -> int:

        ans = 0

        dic = {}
        for idx, name in enumerate(nums):
            a = nums[idx-1] if idx > 0 else 0
            nums[idx] = (name + a) % k
            dic[nums[idx]] = dic.get(nums[idx], 0) + 1

        ans += dic.get(0, 0)
        for i in list(dic.values()):
            if i > 1:
                ans += i * (i-1) // 2

        return ans
```

# **1926**

그래프 BFS

```python
class Solution:
    def nearestExit(self, maze: List[List[str]], e: List[int]) -> int:

        m = len(maze[0])
        n = len(maze)

        visited = [[-1 for _ in range(m)] for _ in range(n)]

        from collections import deque

        queue = deque()
        queue.append((e[1], e[0], 0))
        visited[e[0]][e[1]] = 0

        dx = [1, -1, 0, 0]
        dy = [0, 0, 1, -1]
        ans = -1

        while queue:
            x, y, cnt = queue.popleft()

            if (x == 0 or x == m-1 or y == 0 or y == n-1) and cnt > 0:
                ans = cnt
                break

            for i in range(4):
                nx = x + dx[i]
                ny = y + dy[i]

                if 0<= nx < m and 0<= ny < n and visited[ny][nx] == -1 and maze[ny][nx] == ".":
                    visited[ny][nx] = cnt + 1
                    queue.append((nx, ny, cnt + 1))

        # print(visited)
        # print(ans)
        return ans
```

```python
class Solution:
    def relativeSortArray(self, arr1: List[int], arr2: List[int]) -> List[int]:

        dic = {}
        remain = []
        ans = []

        for i in range(len(arr1)):
            if arr1[i] in arr2:
                dic[arr1[i]] = dic.get(arr1[i], 0) + 1
            else:
                remain.append(arr1[i])

        remain = sorted(remain, key=lambda x: x)

        for i in range(len(arr2)):
            a = arr2[i]

            while dic[a] > 0:
                ans.append(a)
                dic[a] -= 1

            del dic[a]

        return ans + remain
```

```python
class Solution:
    def findMaximizedCapital(self, k: int, W: int, Profits: List[int], Capital: List[int]) -> int:

        projects = [(Capital[i], Profits[i]) for i in range(len(Profits))]
        projects.sort()

        max_heap = []
        i = 0

        for _ in range(k):
            while i < len(projects) and projects[i][0] <= W:
                heapq.heappush(max_heap, -projects[i][1])
                i += 1

            if not max_heap:
                break

            W += -heapq.heappop(max_heap)

        return W
```

```python
class Solution:
    def minPatches(self, nums: List[int], n: int) -> int:
        a, ans, idx = 0, 0, 0

        while a < n:
            if idx < len(nums) and nums[idx] <= a + 1:
                a += nums[idx]
                idx += 1
            else:
                ans += 1
                a = 2*a + 1

        return ans
```

**왜 2 * reach + 1인가?**

reach + 1을 추가하면 reach를 2 * reach + 1로 확장할 수 있습니다.

이는:

reach + 1을 추가하기 전에는 1부터 reach까지 모든 숫자를 만들 수 있었습니다.
reach + 1을 추가함으로써 이제는:
이전의 모든 합 (1부터 reach까지)
그리고 reach + 1부터 2 * reach + 1까지의 모든 합을 만들 수 있습니다.

**예시**
예를 들어, nums = [1, 3]이고 n = 6일 때:

초기 reach = 0.
1 <= 1이므로, 1을 사용하여 reach를 1로 확장합니다.
3 <= 2가 아니므로, 패치 2를 추가합니다. 이제 reach = 2 * 1 + 1 = 3.
3 <= 4이므로, 3을 사용하여 reach를 6으로 확장합니다.
이제 reach >= 6이므로 종료하고 ans = 1을 반환합니다.

```python
class Solution:
    def judgeSquareSum(self, c: int) -> bool:

        l, r = 0, int(math.sqrt(c))

        while l <= r:
            a2 = l*l
            b2 = r*r

            if a2 + b2 == c:
                return True
            elif a2 + b2 < c:
                l += 1
            else:
                r -= 1
        return False
```

# **1482**

```python
class Solution:
    def minDays(self, A: List[int], m: int, k: int) -> int:
        if m * k > len(A): return -1
        left, right = 1, max(A)

        while left < right:
            mid = (left + right) // 2
            flow = bouq = 0

            for a in A:
                flow = 0 if a > mid else flow + 1
                if flow >= k:
                    flow = 0
                    bouq += 1
                    if bouq == m: break

            if bouq == m:
                right = mid
            else:
                left = mid + 1

        return left
```

- 문제를 결정 문제(yes/no 문제)로 변환할 수 있기 때문에 이분 탐색을 사용할 수 있습니다

# **1552**

이분 탐색

```python
class Solution:
    def maxDistance(self, position: List[int], m: int) -> int:

        position.sort()

        l = 1
        r = position[-1] - position[0]
        ans = -1

        while l <= r:
            mid = (l + r) // 2
            cnt, last = 1, position[0]

            for i in range(1, len(position)):
                if position[i] - last >= mid:
                    cnt += 1
                    last = position[i]
                    if cnt == m:
                        break
            if cnt == m:
                ans = mid
                l = mid + 1
            else:
                r = mid - 1

        return ans
```

```python
class Solution:
    def maxSatisfied(self, customers: List[int], grumpy: List[int], minutes: int) -> int:

        n = len(customers)

        window = 0
        for i in range(n):
            if i < minutes:
                window += customers[i]
            else:
                window += customers[i] if grumpy[i] == 0 else 0

        ans = window
        for i in range(1, n-minutes+1):
            a = customers[i+minutes-1] if grumpy[i+minutes-1] == 1 else 0
            b = customers[i-1] if grumpy[i-1] == 1 else 0
            window += a - b
            ans = max(window, ans)

        return ans
```

# **1438**

```python
class Solution:
    def longestSubarray(self, nums: List[int], limit: int) -> int:
        from collections import deque

        a = deque()     #  최소
        b = deque()     #  최대
        n = len(nums)

        l = r = 0
        ans = 0
        while r < len(nums):
            while a and nums[r] <= nums[a[-1]]:
                a.pop()
            while b and nums[r] >= nums[b[-1]]:
                b.pop()
            a.append(r)
            b.append(r)

            while nums[b[0]] - nums[a[0]] > limit:
                l += 1
                if l > a[0]:
                    a.popleft()
                if l > b[0]:
                    b.popleft()

            ans = max(ans, r- l + 1)
            r += 1

        return ans
```

```python
class Solution:
    def minKBitFlips(self, nums: List[int], k: int) -> int:
        n = len(nums)
        flipped = 0
        res = 0
        isFlipped = [0] * n

        for i in range(n):
            if i>=k:
                flipped ^= isFlipped[i-k]
            if flipped == nums[i]:
                if i+k > n:
                    return -1

                isFlipped[i] = 1
                flipped ^= 1
                res += 1

        return res
```

```python
class Solution:
    def getAncestors(self, n: int, edges: List[List[int]]) -> List[List[int]]:
        arr = defaultdict(list)
        ans = [[] for _ in range(n)]

        for edge in edges:
            arr[edge[0]] += [edge[1]]

        def dfs(x, idx):
            for i in arr[idx]:
                if ans[i] and ans[i][-1] == x: continue
                ans[i].append(x)
                dfs(x, i)

        for i in range(n):
            dfs(i, i)

        return ans
```

```python
class Solution:
    def passThePillow(self, n: int, time: int) -> int:

        i = 0
        cnt = 1
        while i < time:

            if (i // (n-1)) % 2 == 0:
                cnt += 1
            else:
                cnt -= 1
            i += 1

        return cnt
```

# **1518**

구현

```python
class Solution:
    def numWaterBottles(self, numBottles: int, numExchange: int) -> int:

        sum = numBottles

        while numBottles >= numExchange:
            cnt = numBottles // numExchange
            numBottles = numBottles % numExchange
            numBottles += cnt
            sum += cnt

        return sum
```
