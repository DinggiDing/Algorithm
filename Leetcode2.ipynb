# **238**

누적합

```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:

        ans = [1]*len(nums)
        wer = [1]*len(nums)
        answer = [0] * len(nums)

        # From LEFT'
        ans[0] = nums[0]
        for i in range(1, len(nums)):
            ans[i] = ans[i-1] * nums[i]

        # From RIGHT
        wer[-1] = nums[-1]
        for i in range(len(nums)-2, -1, -1):
            wer[i] = wer[i+1] * nums[i]

        answer[0] = wer[1]
        answer[-1] = ans[-2]
        for i in range(1, len(nums)-1):
            answer[i] = ans[i-1] * wer[i+1]

        return answer
```

# **1002**

구현

```python
class Solution:
    def commonChars(self, words: List[str]) -> List[str]:

        ans = []
        win = list(words[0])

        for word in words:
            for i in range(len(word)):
                if word[i] in win:
                    ans.append(word[i])
                    # print(win[win.index(word[i])])
                    del win[win.index(word[i])]

            win = ans
            ans = []

        return win
```
