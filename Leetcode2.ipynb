# **238**

누적합

```python
class Solution:
    def productExceptSelf(self, nums: List[int]) -> List[int]:

        ans = [1]*len(nums)
        wer = [1]*len(nums)
        answer = [0] * len(nums)

        # From LEFT'
        ans[0] = nums[0]
        for i in range(1, len(nums)):
            ans[i] = ans[i-1] * nums[i]

        # From RIGHT
        wer[-1] = nums[-1]
        for i in range(len(nums)-2, -1, -1):
            wer[i] = wer[i+1] * nums[i]

        answer[0] = wer[1]
        answer[-1] = ans[-2]
        for i in range(1, len(nums)-1):
            answer[i] = ans[i-1] * wer[i+1]

        return answer
```

# **1002**

```python
class Solution:
    def commonChars(self, words: List[str]) -> List[str]:

        ans = []
        win = list(words[0])

        for word in words:
            for i in range(len(word)):
                if word[i] in win:
                    ans.append(word[i])
                    # print(win[win.index(word[i])])
                    del win[win.index(word[i])]

            win = ans
            ans = []

        return win
```

```python
class Solution:
    def isNStraightHand(self, hand: List[int], groupSize: int) -> bool:

        c = collections.Counter(hand)                       # 수 세기

        for i in sorted(c):
            if c[i] > 0:
                for j in range(groupSize)[::-1]:            # = (2, -1, -1)
                    c[i+j] -= c[i]
                    if c[i+j]<0:
                        return False

        return True
```

```python
class Solution:
    def closeStrings(self, word1: str, word2: str) -> bool:

        w1 = {}
        w2 = {}

        for i in range(len(word1)):
            w1[word1[i]] = w1.get(word1[i], 0) + 1

        for i in range(len(word2)):
            w2[word2[i]] = w2.get(word2[i], 0) + 1

        a1 = sorted(w1.values())
        a2 = sorted(w2.values())
        b1 = sorted(w1.keys())
        b2 = sorted(w2.keys())

        # print(a1, b1)
        # print(a2, b2)

        if a1 == a2 and b1 == b2:
            return True
        else:
            return False
```

# **648**

구현

```python
class Solution:
    def replaceWords(self, dic: List[str], sentence: str) -> str:

        sent = sentence.split(' ')

        di = sorted(dic, key=lambda x: len(x), reverse=False)

        for i in range(len(sent)):
            for j in di:

                if j == sent[i][0:len(j)]:
                    sent[i] = j
                    break

        return ' '.join(sent)
```

```python
class Solution:
    def removeStars(self, s: str) -> str:
        from collections import deque

        queue = deque()

        for i in range(len(s)):
            if s[i] == "*":
                queue.pop()
            else:
                queue.append(s[i])

        # print(queue)

        return ''.join(queue)
```

# **523**

누적합

```python
class Solution:
    def checkSubarraySum(self, nums: List[int], k: int) -> bool:

        # prefix sum
        seen, cur = {0: -1}, 0
        for i, a in enumerate(nums):
            cur = (cur + a) % k if k else cur + a
            if i - seen.setdefault(cur, i) > 1: return True
        return False
```

나머지를 저장할 때, 이전에 같은 나머지가 있었다면 그 나머지 사이에 있는 부분 배열이 k의 배수임을 의미합니다.



> 구체적인 계산 과정


첫 번째 원소: 23 -> 부분 합: 23, 나머지: 23 % 6 = 5

두 번째 원소: 2 -> 부분 합: 25, 나머지: 25 % 6 = 1

세 번째 원소: 4 -> 부분 합: 29, 나머지: 29 % 6 = 5

여기서 나머지가 다시 5가 나왔습니다. 처음에 5가 나온 부분 합은 23이었고, 이번에 29입니다.

23부터 29까지의 부분 배열 합은 6이고, 이는 k의 배수입니다.

# **974**

누적합 , 523의 응용

```python
class Solution:
    def subarraysDivByK(self, nums: List[int], k: int) -> int:

        ans = 0

        dic = {}
        for idx, name in enumerate(nums):
            a = nums[idx-1] if idx > 0 else 0
            nums[idx] = (name + a) % k
            dic[nums[idx]] = dic.get(nums[idx], 0) + 1

        ans += dic.get(0, 0)
        for i in list(dic.values()):
            if i > 1:
                ans += i * (i-1) // 2

        return ans
```

2중 반복문은 시간 초과

딕셔너리로
    # 0 이상의 나머지가 2개 이상일때 → 4C2 = 4*3 / 2 = 6
    # 0 의 나머지일 때 → n 개

```python
# 최적화 솔루션

ef subarraysDivByK(self, A, K):
    res = 0
    prefix = 0
    count = [1] + [0] * K
    for a in A:
        prefix = (prefix + a) % K
        res += count[prefix]
        count[prefix] += 1
    return res
```

```python
# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def leafSimilar(self, root1: Optional[TreeNode], root2: Optional[TreeNode]) -> bool:

        a = self.dfs(root1)
        b = self.dfs(root2)

        return True if a == b else False


    def dfs(self, root):
        if not root:
            return []
        if root.left == None and root.right == None:
            return [root.val]

        return self.dfs(root.left) + self.dfs(root.right)


```

# **1926**

그래프 BFS

```python
class Solution:
    def nearestExit(self, maze: List[List[str]], e: List[int]) -> int:

        m = len(maze[0])
        n = len(maze)

        visited = [[-1 for _ in range(m)] for _ in range(n)]

        from collections import deque

        queue = deque()
        queue.append((e[1], e[0], 0))
        visited[e[0]][e[1]] = 0

        dx = [1, -1, 0, 0]
        dy = [0, 0, 1, -1]
        ans = -1

        while queue:
            x, y, cnt = queue.popleft()

            if (x == 0 or x == m-1 or y == 0 or y == n-1) and cnt > 0:
                ans = cnt
                break

            for i in range(4):
                nx = x + dx[i]
                ny = y + dy[i]

                if 0<= nx < m and 0<= ny < n and visited[ny][nx] == -1 and maze[ny][nx] == ".":
                    visited[ny][nx] = cnt + 1
                    queue.append((nx, ny, cnt + 1))

        # print(visited)
        # print(ans)
        return ans
```

# **1122**

딕셔너리

```python
class Solution:
    def relativeSortArray(self, arr1: List[int], arr2: List[int]) -> List[int]:

        dic = {}
        remain = []
        ans = []

        for i in range(len(arr1)):
            if arr1[i] in arr2:
                dic[arr1[i]] = dic.get(arr1[i], 0) + 1
            else:
                remain.append(arr1[i])

        remain = sorted(remain, key=lambda x: x)

        for i in range(len(arr2)):
            a = arr2[i]

            while dic[a] > 0:
                ans.append(a)
                dic[a] -= 1

            del dic[a]

        return ans + remain
```

# **2095**

```python
# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def deleteMiddle(self, head: Optional[ListNode]) -> Optional[ListNode]:

        if head.next == None: return None

        n = 0

        b = ListNode(0)
        b.next = head
        c = head

        while c:
            c = c.next
            n += 1

        cnt = n // 2 if n % 2 == 0 else n // 2 + 1
        if n % 2 == 0:
            d = head
        else:
            d = ListNode(0)
            d.next = head

        n2  = 0
        # print(n, cnt)
        while d:
            # print(n2, d)
            if n2 == cnt-1:
                if d.next == None:
                    d = None
                else:
                    # print("======")
                    d.next = d.next.next
                    d = d.next
                    n2 += 1
            else:
                d = d.next
                n2 += 1

        return b.next
```

# **724**

누적합

```python
class Solution:
    def pivotIndex(self, nums: List[int]) -> int:

        # 누적합 : 왼쪽, 오른쪽으로 각각 시작하는 누적합 구해서 left = right - 2인 것 찾기
        # 투포인터

        n = len(nums)
        a = [0 for _ in range(n)]
        b = [0 for _ in range(n)]

        a[0] = nums[0]
        b[-1] = nums[-1]
        for i in range(1, len(nums)):
            a[i] = nums[i] + a[i-1]
            b[n-i-1] = nums[n-i-1] + b[n-i]

        a.insert(0, 0)
        a.append(0)
        b.insert(0, 0)
        b.append(0)

        ans = -1
        for i in range(n):
            if a[i] == b[i+2]:
                ans = i
                break

        return ans
```

# **2037**

```python
class Solution:
    def minMovesToSeat(self, seats: List[int], students: List[int]) -> int:

        seats = sorted(seats, key=lambda x: x)
        students = sorted(students, key=lambda x: x)

        ans = 0
        for i in range(len(seats)):
            ans += abs(seats[i] - students[i])

        return ans
```
