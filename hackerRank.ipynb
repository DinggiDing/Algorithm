# **Climbing the Leaderboard**

```python
def climbingLeaderboard(ranked, player):
    # Write your code here

    st = sorted(set(ranked), reverse=True)
    ans = []
    l = len(st)
    for p in player:
        while l > 0 and p >= st[l - 1]:
            l -= 1
        ans.append(l+1)
    return ans
```

# **Connected Cells in a Grid**

DFS

```python
def connectedCell(matrix):
    # Write your code here

    m = len(matrix[0])
    n = len(matrix)

    visited = [[0 for _ in range(m)] for _ in range(n)]
    dx = [1, -1, 0, 0, 1, 1, -1, -1]
    dy = [0, 0, -1, 1, 1, -1, -1, 1]
    cnt = [1]

    def dfs(x, y):

        for i in range(8):
            nx = x + dx[i]
            ny = y + dy[i]

            if 0<= nx <m and 0<= ny <n and matrix[ny][nx] == 1 and visited[ny][nx] == 0:
                cnt[0] = cnt[0] + 1
                visited[ny][nx] = cnt[0]
                dfs(nx, ny)


    for i in range(n):
        for j in range(m):
            if matrix[i][j] == 1 and visited[i][j] == 0:
                cnt[0] = 1
                visited[i][j] = 1
                dfs(j, i)

    # for i in visited:
    #     print(i)

    return max(map(max, visited))
```

# **Non-Divisible Subset**

```python
def nonDivisibleSubset(k, s):
    # Write your code here

    remainder_counts = [0] * k

    for num in s:
        remainder_counts[num % k] += 1

    max_subset_size = 0

    if remainder_counts[0] > 0:
        max_subset_size += 1

    for i in range(1, (k // 2) + 1):
        if i != k - i:
            max_subset_size += max(remainder_counts[i], remainder_counts[k - i])
        else:
            max_subset_size += min(remainder_counts[i], 1)

    return max_subset_size
```

# **Bigger is Greater**
