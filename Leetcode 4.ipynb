```python
class Solution:
    def lemonadeChange(self, bills: List[int]) -> bool:
        five = 0
        ten = 0
        ans = True

        for i in bills:
            if i == 5:
                five += 1
            elif i == 10:
                ten += 1
                five -= 1
            else:
                if ten > 0:
                    ten -= 1
                else:
                    five -= 2
                five -= 1
            if five < 0 or ten < 0:
                ans = False

        return ans
```
