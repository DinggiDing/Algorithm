2차원 배열 DP 합

```python
class Solution:
    def uniquePaths(self, m: int, n: int) -> int:
        dp = [[0 for _ in range(m)] for _ in range(n)]

        for i in range(n):
            dp[i][0] = 1
            for j in range(1, m):
                if i == 0:
                    dp[i][j] = 1
                else:
                    dp[i][j] = dp[i-1][j] + dp[i][j-1]

        return dp[n-1][m-1]
```

# **2140**

```python
class Solution:
    def mostPoints(self, questions: List[List[int]]) -> int:

        n = len(questions)
        DP = [0 for _ in range(n+1)]
        NN = {}
        cnt_point = 0

        for i in range(1, n+1):
            a = NN.get(i, 0)
            if a != 0:
                cnt_point = max(DP[a], cnt_point)

            DP[i] = cnt_point + questions[i-1][0]

            if i + questions[i-1][1] + 1 <= n:
                if DP[NN.get(i + questions[i-1][1] + 1, 0)] < DP[i]:
                    NN[i + questions[i-1][1] + 1] = i

        # print(DP)
        # print(NN)
        return max(DP)
```

시간 최적화 진행

**뒤에서부터 DP를 선택하는 근거**

"뒤에서부터 DP를 사용하는 이유"를 일반적인 DP 문제와 연관 지어 설명하면 다음과 같습니다:

&nbsp;

- **미래의 상태가 현재의 결정을 좌우하는 경우**: 현재 상태에서 가능한 결정들이 미래의 상태에 따라 달라질 때 뒤에서부터 접근하는 것이 유리합니다. 즉, 미래 상태에서 최적 결정을 알고 있다면 현재 상태에서 그 결정을 쉽게 내릴 수 있습니다.

- **종료 조건이 명확한 경우**: 뒤에서부터 접근하면 종료 조건(예: 배열의 끝, 문자열의 끝 등)이 명확하게 주어질 때가 많습니다. 이렇게 하면 자연스럽게 DP 테이블을 채우는 방향이 결정됩니다.

- **문제의 성질에 따라**: 예를 들어, 점프 게임이나 연속적인 결정 문제에서는 뒤에서부터 접근하여 각 위치에서 가능한 점프나 결정을 고려하는 것이 더 직관적일 수 있습니다.



```python
class Solution:
    def mostPoints(self, questions: List[List[int]]) -> int:

        n = len(questions)
        dp = [0] * (n + 1)  # dp 배열을 0으로 초기화 (퀘스트의 수보다 하나 더 크게)

        # 뒤에서부터 DP 배열을 채웁니다.
        for i in range(n - 1, -1, -1):
            points, brainpower = questions[i]
            next_question = i + brainpower + 1
            if next_question < n:
                dp[i] = max(points + dp[next_question], dp[i + 1])
            else:
                dp[i] = max(points, dp[i + 1])

        return dp[0]
```

```python
class Solution:
    def numSquares(self, n: int) -> int:

        dp = [n] * (n+1)
        dp[0] = 0

        c = 1
        while c * c <= n:
            s = c*c
            for i in range(s, n+1):
                dp[i] = min(dp[i-s]+1, dp[i])
            c += 1
        return dp[n]
```

# **72**

```python
class Solution:
    def minDistance(self, word1: str, word2: str) -> int:

        n = len(word1)
        m = len(word2)
        dp = [[0 for _ in range(n+1)] for _ in range(m+1)]

        for i in range(n+1):
            dp[0][i] = i
        for i in range(m+1):
            dp[i][0] = i


        for i in range(1, m+1):
            for j in range(1, n+1):
                if word1[j-1] == word2[i-1]:
                    dp[i][j] = dp[i-1][j-1]  # no cost
                else:
                    dp[i][j] = min(dp[i-1][j-1], dp[i-1][j], dp[i][j-1]) + 1

        return dp[m][n]
```

# **673**

```python
class Solution:
    def findNumberOfLIS(self, nums: List[int]) -> int:

        n = len(nums)
        if n==1:
            return n
        count = [1] * n                     # 개수
        dp = [1] * n                        # 길이

        for i in range(n):
            for j in range(i):
                if nums[j] < nums[i]:
                    if dp[j] + 1 > dp[i]:
                        dp[i] = dp[j] + 1
                        count[i] = count[j]
                    elif dp[j] + 1 == dp[i]:
                        count[i] += count[j]

        max_len = max(dp)
        return sum(count[i] for i in range(n) if dp[i] == max_len)
```

# **646**

```python
class Solution:
    def findLongestChain(self, pairs: List[List[int]]) -> int:

        n = len(pairs)
        dp = [0] * (n)

        pairs = sorted(pairs, key=lambda x: (x[0], x[1]))

        for i in range(n):
            for j in range(i):
                if pairs[i][0] > pairs[j][1]:
                    dp[i] = max(dp[i], dp[j]+1)

        # print(dp)

        return dp[n-1] +1
```

# **1218**



for i: for j 로 O(n2)로만 생각하였으나 n의 길이가 10^5으로 매우 큼. n^2은 n이 5000정도일때 풀수있음.

연속 배열이 아닌 이전 꺼를 확인하고자 할 때, 딕셔너리를 활용하여 효율적으로 풀 수 있음.



```python
class Solution:
    def longestSubsequence(self, arr: List[int], dif: int) -> int:

        n = len(arr)
        dic = {}

        for i in range(n):
            dic[arr[i]] = dic.get(arr[i]-dif, 0) + 1

        return max(dic.values())
```
