###- 프로세스와 스레드의 차이점, iOS에서의 프로세스와 스레드 관리 방법에 대해 설명해주세요.**

프로세스는 실행 중인 프로그램의 인스턴스로, 각각 독립된 메모리 공간(주소 공간)을 갖습니다. 프로세스 간의 메모리 접근은 원칙적으로 불가능하며, 상호작용을 위해서는 IPC(Inter-Process Communication) 메커니즘을 사용해야 합니다.

스레드는 프로세스 내에서 실행되는 작은 실행 단위. 다중 스레드를 통해 병렬 처리가 가능.

→ GCD : 작업을 비동기적으로 수행하기 위한 API로 큐에 작업을 추가하면 적절한 스레드에서 작업을 실행

###- 메모리 관리 기법 중 iOS에서 사용되는 방식과 그 특징에 대해 설명해주세요.**

ARC : 컴파일러가 객체의 생명주기를 자동으로 관리하는 메모리 관리 방식으로 객체의 참조 수를 추적함.

객체풀링 : 자주 사용되는 객체를 미리 생성해두고 재사용하여 메모리 할당과 해제 비용을 줄이는 기법

&nbsp;

###- iOS에서의 REST API 네트워크 요청 및 응답 처리 방법

URLSession을 사용하여 HTTP 및 기타 네트워크 요청을 처리. JSON과 같은 형식으로 데이터를 디코딩하여 모델 객체로 변환함.

&nbsp;

###- iOS 디바이스의 메모리 제약과 앱 메모리 제한에 대해 설명해주세요.

앱이 사용할 수 있는 메모리가 한정되어 있으며, 이를 초과하면 시스템은 앱을 종료시킵니다.

이를 해결하기 위해 크고 메모리 사용량이 많은 이미지를 사용할 때는 필요할 때만 로드하고, 사용 후 해제함.

또는 Instruments를 사용하여 메모리 사용을 프로파일링하고, 메모리 누수를 찾아 해결함. 특히 Instruments에는 Allocations(메모리 할당 추적), Leaks(메모리 누수 탐지), Zombies(해제된 객체에 접근하려는 시도 감지)

###- 자주 사용되는 정렬 알고리즘(예: 퀵 정렬, 병합 정렬)의 동작 원리와 시간 복잡도를 설명해주세요.
1. 퀵 정렬 : 분할 정복을 이용합니다. 하나의 pivot을 정하고 작은 것들은 왼쪽, 큰 것들은 오른쪽에 위치하여 분할시켜가며 재귀적으로 수행한 후, 합칩니다. O(nlogn) ~ O(n^2)
2. 병합 정렬 : 분할 정복을 이용합니다. 리스트를 반으로 나눠가면 정렬합니다. O(nlogn)
3. 힙 정렬 : 힙을 이용합니다. O(nlogn)

###- 싱글톤 패턴?

특정 클래스의 인스턴스를 하나만 생성하고, 그 인스턴스에 전역적으로 접근할 수 있도록 하는 패턴. 주로 공유 자원에 접근하거나 로깅, 캐싱 등에 사용됨. 단점으로는 유지보수 어려움, 테스트 어려움 존재

###- 클로저

코드의 재사용성과 모듈성을 높이는 데 유용합니다. 클로저를 사용하면 코드 블록을 변수나 상수처럼 다룰 수 있고, 함수의 매개변수로 전달하거나 반환값으로 사용할 수 있습니다.

&nbsp;


###- 구조체(struct) vs 클래스(class)

구조체에서는 값 타입으로 원본과 복사본이 독립적인 값을 가지고, 간단한 데이터 모델, 스택, 큐와 같은 자료 구조에 사용됩니다. 상속이 필요 없고, 값이 복사될 때 독립적인 복사본을 만드는 경우에 유용합니다.

클래스에서는 참조 타입으로 복사본에 원본의 값이 참조됨. 상속이 필요하거나 참조를 통해 동일한 인스턴스를 공유해야 하는 경우에 사용됩니다. 복잡한 데이터 모델과 상속 계층을 필요로 하는 경우에 적합합니다.

&nbsp;


###- Coredata의 주요 구성 요소

Entity : 하나의 데이터 모델 객체

Attribute : 특정 데이터를 나타내느 필드, 데이터베이스의 column가 비슷한 개념

Relationship : 엔티티 간의 연결, 일대일, 일대다, 다대다 등의 형태 존재

###- 코어 데이터에서 데이터를 가져오는 방법(Fetch Request)에 대해 설명해주세요.

Fetchrequest를 통해 다양한 조건이나 정렬 등을 추가적으로 사용해 데이터를 가져올 수 있으며, 배치 오퍼레이션이나 패치 제한을 두어 메모리 최적화를 할 수 있음.

- 코어 데이터 마이그레이션(Migration)은 언제 필요한가요?

새로운 속성을 추가, 수정, 삭제하거나 데이터 모델 구조가 변경될 경우

- Publisher와 Subscriber의 역할은 무엇인가요?

`Publisher`는 데이터를 생성하고 이벤트를 전달하며, `Subscriber`는 이 데이터를 수신하고 처리합니다. 이러한 관계를 통해 비동기 이벤트를 관리하고 처리하는 강력한 메커니즘을 제공합니다
