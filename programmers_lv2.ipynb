# ***석유시추***

효율성 점수 없음

```python
import numpy as np
from collections import deque

def solution(land):
    answer = 0

    n, m = len(land), len(land[0])
    print(n, m, land[0][1])



    for j in range(m):
      count = 0

      visited = [[False for col in range(m)] for row in range(n)]

      for i in range(n):

        if land[i][j]==1 and visited[i][j]==False:
          # 큐 구현을 위한 deque 라이브러리 활용
          queue = deque([(i, j)])
          # 현재 노드를 방문 처리
          visited[i][j] = True
          count += 1

          # 큐가 완전히 빌 때까지 반복
          while queue:
              # 큐에 삽입된 순서대로 노드 하나 꺼내기
              ii, jj = queue.popleft()

              # 현재 처리 중인 노드에서 방문하지 않은 인접 노드를 모두 큐에 삽입
              dw = [1, 0, -1, 0]
              dh = [0, -1, 0, 1]

              for k in range(4):
                if ii + dw[k] in range(n) and jj + dh[k] in range(m):
                  if not (land[ii + dw[k]][jj + dh[k]] == 0 or visited[ii + dw[k]][jj + dh[k]]):
                      queue.append((ii + dw[k], jj+dh[k]))
                      visited[ii + dw[k]][jj+ dh[k]] = True
                      count += 1

        if answer < count:
          answer = count

    return answer
```

# **연속된 부분 수열의 합**

```python
def solution(sequence, k):
    answer = []

    n = len(sequence)
    a1 = 0
    a2 = 0

    for i in range(n-1, -1, -1):
      sum = k
      sum -= sequence[i]
      if sum==0:
        if a1>0:
          a1 = i
          a2 = i
        else:
          a1 = i
          a2 = i
          break
      for j in range(i-1, -1, -1):
        if sum!=0:
          sum -= sequence[j]
        if sum==0:
          if a1>0:
            a1 = j
            a2 = i
          else:
            a1 = j
            a2 = i
          break

    answer.append(a1)
    answer.append(a2)


    return answer

seq = [1, 2, 3, 4, 5, 5]
k = 5
solution(seq, k)
```

# **혼자 놀기의 달인**

https://school.programmers.co.kr/learn/courses/30/lessons/131130

```python
def solution(cards):
    answer = 0

    n = len(cards)
    cards_sw = [False for i in range(n)]   # cards 요소에 접근했는지 여부 배열
    answer_arr = []                         # 정답 배열 추후 정렬해서 1,2 값 곱


    for i in range(n):

      if not cards_sw[i]:
        cnt = 0
        a = i
        while True:
          b = cards[a]
          cards_sw[a]=True
          cnt += 1
          if i==b-1:
            answer_arr.append(cnt)
            break
          else:
            a = b-1

    print(answer_arr)

    if len(answer_arr)>1:
      answer_arr.sort(reverse=True)
      answer = answer_arr[0]*answer_arr[1]
    else:
      answer = 0

    return answer

cards = [8,6,3,7,2,5,1,4]
solution(cards)
```

# **연속 부분 수열 합의 개수**

```python
def solution(elements):
    answer = 0

    n = len(elements)
    answer_set = set()

    for i in range(n):
      sum = 0
      for j in range(n):
        a = j+i

        if a>=n:
          a = a-n

        sum += elements[a]
        answer_set.add(sum)

    answer = len(answer_set)

    return answer


elements = [7,9,1,1,4]
solution(elements)
```

# **우박수열 정적분**

https://school.programmers.co.kr/learn/courses/30/lessons/134239

```python
def solution(k, ranges):
    answer = []

    arr = []
    arr.append(k)
    while True:
      if k==1:
        break

      if k%2==0:
        k = k//2
        arr.append(k)
      else:
        k = 3*k+1
        arr.append(k)

    n = len(arr)-1
    for i in range(len(arr)-1):
      arr[i] = (arr[i]+arr[i+1])/2
      if i>0:
        arr[i] += arr[i-1]

    del arr[len(arr)-1]
    arr.insert(0, 0)

    print(n, arr)

    for i in range(len(ranges)):
      if n+ranges[i][1]<ranges[i][0]:
        answer.append(-1.0)
      else:
        answer.append(arr[n+ranges[i][1]] - arr[ranges[i][0]])

    return answer


k = 3
ranges = 	[[0,0], [1,-2], [3,-3], [1, 0]]
solution(k, ranges)
```

# **테이블 해시 함수**

```python
def solution(data, col, row_begin, row_end):
    answer = 0

    n = len(data[0])

    data.sort(key=lambda x : (x[col-1], -x[0]))

    sum2 = 0
    for i in range(row_begin-1, row_end):
      sum = 0

      for j in range(n):
        sum += data[i][j]%(i+1)

      if i!=row_begin-1:
        sum2 = sum2 ^ sum
      else:
        sum2 = sum


    answer = sum2
    return answer


data = [[2,2,6],[1,5,10],[4,2,9],[3,8,3]]
col = 2
row_begin = 2
row_end = 3
solution(data, col, row_begin, row_end)
```

# **이모티콘 할인행사** : *다시 풀기*

```python
from itertools import product

def solution(users, emoticons):
    emg_len = len(emoticons)
    answer = [0, 0]

    for p in product(range(10, 41, 10), repeat=emg_len):        #조합
        counts, money = 0, 0
        for i in range(len(users)):
            user_discount = users[i][0]
            total = 0

            for j in range(len(p)):
                if user_discount <= p[j]:
                    total += emoticons[j] * (100 - p[j]) * 0.01

            if total >= users[i][1]:
                counts += 1

            else:
                money += total

        if answer[0] > counts:
            continue

        if answer[0] == counts and answer[1] > money:
            continue

        answer = [counts, money]

    return answer


users = [[40, 10000], [25, 10000]]
emoticons = [7000, 9000]
solution(users, emoticons)
```
