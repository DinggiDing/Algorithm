# **1701**

구현

```python
class Solution:
    def averageWaitingTime(self, customers: List[List[int]]) -> float:

        mx = 10 ** 9
        idx = 0
        s = 0
        start = 0
        ongoing = False
        i = 1

        while i < mx:
            if idx == len(customers) and ongoing == False:
                break

            if customers[idx][0] <= i and ongoing == False:
                idx += 1
                ongoing = True

            if ongoing == True:
                start = customers[idx-1][0]
                i += customers[idx-1][1]
                s += i-start
                ongoing = False
                # print(start, i)
            else:
                i += 1

        return s / idx
```

```python
class Solution:
    def averageWaitingTime(self, customers: List[List[int]]) -> float:
        available_at = 0
        total_wait = 0
        for arrival, t in customers:
            available_at = max(available_at, arrival) + t
            total_wait += available_at - arrival

        return total_wait / len(customers)
```

```python
class Solution:
    def minOperations(self, logs: List[str]) -> int:

        depth = 0

        for log in logs:
            lo, g = log.split("/")
            if lo == "..":
                depth -= 1 if depth > 0 else 0
            elif lo == ".":
                depth = depth
            else:
                depth += 1

        return depth
```

# **1190**

```python
class Solution:
    def reverseParentheses(self, s: str) -> str:
        res = ['']
        for c in s:
            if c == '(':
                res.append('')
            elif c == ')':
                res[len(res) - 2] += res.pop()[::-1]
            else:
                res[-1] += c

        print(res)
        return "".join(res)
```

# **1717**

```python
class Solution:
    def maximumGain(self, s: str, x: int, y: int) -> int:
        a_count = 0
        b_count = 0
        lesser = min(x, y)
        result = 0

        for c in s:
            if c > 'b':
                result += min(a_count, b_count) * lesser
                a_count = 0
                b_count = 0
            elif c == 'a':
                if x < y and b_count > 0:
                    b_count -= 1
                    result += y
                else:
                    a_count += 1
            elif c == 'b':
                if x > y and a_count > 0:
                    a_count -= 1
                    result += x
                else:
                    b_count += 1

        result += min(a_count, b_count) * lesser
        return result
```
