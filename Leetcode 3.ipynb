# **1701**

구현

```python
class Solution:
    def averageWaitingTime(self, customers: List[List[int]]) -> float:

        mx = 10 ** 9
        idx = 0
        s = 0
        start = 0
        ongoing = False
        i = 1

        while i < mx:
            if idx == len(customers) and ongoing == False:
                break

            if customers[idx][0] <= i and ongoing == False:
                idx += 1
                ongoing = True

            if ongoing == True:
                start = customers[idx-1][0]
                i += customers[idx-1][1]
                s += i-start
                ongoing = False
                # print(start, i)
            else:
                i += 1

        return s / idx
```

```python
class Solution:
    def averageWaitingTime(self, customers: List[List[int]]) -> float:
        available_at = 0
        total_wait = 0
        for arrival, t in customers:
            available_at = max(available_at, arrival) + t
            total_wait += available_at - arrival

        return total_wait / len(customers)
```

# **1190**

```python
class Solution:
    def reverseParentheses(self, s: str) -> str:
        res = ['']
        for c in s:
            if c == '(':
                res.append('')
            elif c == ')':
                res[len(res) - 2] += res.pop()[::-1]
            else:
                res[-1] += c

        print(res)
        return "".join(res)
```

# **1717**

```python
class Solution:
    def maximumGain(self, s: str, x: int, y: int) -> int:
        a_count = 0
        b_count = 0
        lesser = min(x, y)
        result = 0

        for c in s:
            if c > 'b':
                result += min(a_count, b_count) * lesser
                a_count = 0
                b_count = 0
            elif c == 'a':
                if x < y and b_count > 0:
                    b_count -= 1
                    result += y
                else:
                    a_count += 1
            elif c == 'b':
                if x > y and a_count > 0:
                    a_count -= 1
                    result += x
                else:
                    b_count += 1

        result += min(a_count, b_count) * lesser
        return result
```

# **2751**

구현

인덱스를 위한 배열을 따로 생성하여 정렬해도 입력 순서 그대로 출력 가능

```python
class Solution:
    def survivedRobotsHealths(self, positions: List[int], h: List[int], directions: str) -> List[int]:
        n = len(positions)
        ind = sorted(range(n), key=positions.__getitem__)
        stack = []
        for i in ind:
            if directions[i] == 'R':
                stack.append(i)
                continue
            while stack and h[i] > 0:
                if h[stack[-1]] < h[i]:
                    h[stack.pop()] = 0
                    h[i] -= 1
                elif h[stack[-1]] > h[i]:
                    h[stack[-1]] -= 1
                    h[i] = 0
                else:
                    h[stack.pop()] = 0
                    h[i] = 0

        return [v for v in h if v > 0]
```
