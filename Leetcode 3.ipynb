# **1701**

구현

```python
class Solution:
    def averageWaitingTime(self, customers: List[List[int]]) -> float:

        mx = 10 ** 9
        idx = 0
        s = 0
        start = 0
        ongoing = False
        i = 1

        while i < mx:
            if idx == len(customers) and ongoing == False:
                break

            if customers[idx][0] <= i and ongoing == False:
                idx += 1
                ongoing = True

            if ongoing == True:
                start = customers[idx-1][0]
                i += customers[idx-1][1]
                s += i-start
                ongoing = False
                # print(start, i)
            else:
                i += 1

        return s / idx
```

```python
class Solution:
    def averageWaitingTime(self, customers: List[List[int]]) -> float:
        available_at = 0
        total_wait = 0
        for arrival, t in customers:
            available_at = max(available_at, arrival) + t
            total_wait += available_at - arrival

        return total_wait / len(customers)
```

```python
class Solution:
    def reverseParentheses(self, s: str) -> str:
        res = ['']
        for c in s:
            if c == '(':
                res.append('')
            elif c == ')':
                res[len(res) - 2] += res.pop()[::-1]
            else:
                res[-1] += c

        print(res)
        return "".join(res)
```

# **1717**

```python
class Solution:
    def maximumGain(self, s: str, x: int, y: int) -> int:
        a_count = 0
        b_count = 0
        lesser = min(x, y)
        result = 0

        for c in s:
            if c > 'b':
                result += min(a_count, b_count) * lesser
                a_count = 0
                b_count = 0
            elif c == 'a':
                if x < y and b_count > 0:
                    b_count -= 1
                    result += y
                else:
                    a_count += 1
            elif c == 'b':
                if x > y and a_count > 0:
                    a_count -= 1
                    result += x
                else:
                    b_count += 1

        result += min(a_count, b_count) * lesser
        return result
```

# **2751**

구현

인덱스를 위한 배열을 따로 생성하여 정렬해도 입력 순서 그대로 출력 가능

```python
class Solution:
    def survivedRobotsHealths(self, positions: List[int], h: List[int], directions: str) -> List[int]:
        n = len(positions)
        ind = sorted(range(n), key=positions.__getitem__)
        stack = []
        for i in ind:
            if directions[i] == 'R':
                stack.append(i)
                continue
            while stack and h[i] > 0:
                if h[stack[-1]] < h[i]:
                    h[stack.pop()] = 0
                    h[i] -= 1
                elif h[stack[-1]] > h[i]:
                    h[stack[-1]] -= 1
                    h[i] = 0
                else:
                    h[stack.pop()] = 0
                    h[i] = 0

        return [v for v in h if v > 0]
```

# **726**

구현(Hard)

```python
class Solution:
    def countOfAtoms(self, formula: str) -> str:

        dic = collections.defaultdict(int)
        coeff = 1
        stack = []
        elem = ""
        cnt = 0
        i = 0           # '112'를 숫자화하기 위해

        for s in formula[::-1]:
            if s.isdigit():
                cnt += int(s) * (10 ** i)
                i += 1
            elif s == ")":
                if cnt == 0:
                    cnt += 1
                stack.append(cnt)
                coeff *= cnt
                i = cnt = 0
            elif s == "(":
                coeff //= stack.pop()
                i = cnt = 0
            elif s.isupper():
                elem += s
                dic[elem[::-1]] += (cnt or 1)*coeff
                elem = ""
                i = cnt = 0
            elif s.islower():
                elem += s

        return "".join(k + str(v > 1 and v or "") for k, v in sorted(dic.items()))
```

# **1380**

```python
class Solution:
    def luckyNumbers (self, matrix: List[List[int]]) -> List[int]:
        m = len(matrix[0])
        n = len(matrix)
        ans = 0
        check = False

        for i in range(n):
            a = 1000000
            b = -1
            for j in range(m):
                if matrix[i][j] < a:
                    b = j
                    a = matrix[i][j]

            c = False

            for k in range(n):
                if matrix[k][b] > a:
                    c = True

            if c == False:
                ans = a
                check = True

        return [ans] if check else []
```

# **1605**

```python
class Solution:
    def restoreMatrix(self, row: List[int], col: List[int]) -> List[List[int]]:
        m, n = len(row), len(col)
        A = [[0] * n for i in range(m)]
        for i in range(m):
            for j in range(n):
                A[i][j] = min(row[i], col[j])
                row[i] -= A[i][j]
                col[j] -= A[i][j]
        return A
```

# **2392**

위상 정렬 (Topological Sort)

```python
class Solution:
    def buildMatrix(self, k: int, R: List[List[int]], C: List[List[int]]) -> List[List[int]]:

        def helper(A):
            nxt, indegree = [set() for _ in range(k)], [0] * k
            dq, ans = collections.deque(), []
            A = set([tuple(a) for a in A])
            for i, j in A:
                nxt[i - 1].add(j - 1)
                indegree[j - 1] += 1
            for i in range(k):
                if indegree[i] == 0:
                    dq.append(i)
            while dq:
                cur = dq.popleft()
                ans.append(cur)
                for cand in nxt[cur]:
                    indegree[cand] -= 1
                    if indegree[cand] == 0:
                        dq.append(cand)
            return ans if len(ans) == k else []

        ans1, ans2 = helper(R), helper(C)
        if not ans1 or not ans2:
            return []

        A = [[0] * k for _ in range(k)]
        for i in range(k):
            A[ans1.index(i)][ans2.index(i)] = i + 1
        return A
```
