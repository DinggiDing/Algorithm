# **1701**

```python
class Solution:
    def averageWaitingTime(self, customers: List[List[int]]) -> float:

        mx = 10 ** 9
        idx = 0
        s = 0
        start = 0
        ongoing = False
        i = 1

        while i < mx:
            if idx == len(customers) and ongoing == False:
                break

            if customers[idx][0] <= i and ongoing == False:
                idx += 1
                ongoing = True

            if ongoing == True:
                start = customers[idx-1][0]
                i += customers[idx-1][1]
                s += i-start
                ongoing = False
                # print(start, i)
            else:
                i += 1

        return s / idx
```

```python
class Solution:
    def averageWaitingTime(self, customers: List[List[int]]) -> float:
        available_at = 0
        total_wait = 0
        for arrival, t in customers:
            available_at = max(available_at, arrival) + t
            total_wait += available_at - arrival

        return total_wait / len(customers)
```

```python
class Solution:
    def maximumGain(self, s: str, x: int, y: int) -> int:
        a_count = 0
        b_count = 0
        lesser = min(x, y)
        result = 0

        for c in s:
            if c > 'b':
                result += min(a_count, b_count) * lesser
                a_count = 0
                b_count = 0
            elif c == 'a':
                if x < y and b_count > 0:
                    b_count -= 1
                    result += y
                else:
                    a_count += 1
            elif c == 'b':
                if x > y and a_count > 0:
                    a_count -= 1
                    result += x
                else:
                    b_count += 1

        result += min(a_count, b_count) * lesser
        return result
```

# **2751**

구현

인덱스를 위한 배열을 따로 생성하여 정렬해도 입력 순서 그대로 출력 가능

```python
class Solution:
    def survivedRobotsHealths(self, positions: List[int], h: List[int], directions: str) -> List[int]:
        n = len(positions)
        ind = sorted(range(n), key=positions.__getitem__)
        stack = []
        for i in ind:
            if directions[i] == 'R':
                stack.append(i)
                continue
            while stack and h[i] > 0:
                if h[stack[-1]] < h[i]:
                    h[stack.pop()] = 0
                    h[i] -= 1
                elif h[stack[-1]] > h[i]:
                    h[stack[-1]] -= 1
                    h[i] = 0
                else:
                    h[stack.pop()] = 0
                    h[i] = 0

        return [v for v in h if v > 0]
```

# **726**

구현(Hard)

```python
class Solution:
    def countOfAtoms(self, formula: str) -> str:

        dic = collections.defaultdict(int)
        coeff = 1
        stack = []
        elem = ""
        cnt = 0
        i = 0           # '112'를 숫자화하기 위해

        for s in formula[::-1]:
            if s.isdigit():
                cnt += int(s) * (10 ** i)
                i += 1
            elif s == ")":
                if cnt == 0:
                    cnt += 1
                stack.append(cnt)
                coeff *= cnt
                i = cnt = 0
            elif s == "(":
                coeff //= stack.pop()
                i = cnt = 0
            elif s.isupper():
                elem += s
                dic[elem[::-1]] += (cnt or 1)*coeff
                elem = ""
                i = cnt = 0
            elif s.islower():
                elem += s

        return "".join(k + str(v > 1 and v or "") for k, v in sorted(dic.items()))
```

# **2392**

위상 정렬 (Topological Sort)

```python
class Solution:
    def buildMatrix(self, k: int, R: List[List[int]], C: List[List[int]]) -> List[List[int]]:

        def helper(A):
            nxt, indegree = [set() for _ in range(k)], [0] * k
            dq, ans = collections.deque(), []
            A = set([tuple(a) for a in A])
            for i, j in A:
                nxt[i - 1].add(j - 1)
                indegree[j - 1] += 1
            for i in range(k):
                if indegree[i] == 0:
                    dq.append(i)
            while dq:
                cur = dq.popleft()
                ans.append(cur)
                for cand in nxt[cur]:
                    indegree[cand] -= 1
                    if indegree[cand] == 0:
                        dq.append(cand)
            return ans if len(ans) == k else []

        ans1, ans2 = helper(R), helper(C)
        if not ans1 or not ans2:
            return []

        A = [[0] * k for _ in range(k)]
        for i in range(k):
            A[ans1.index(i)][ans2.index(i)] = i + 1
        return A
```

# **34**

이진탐색

```python
class Solution:
    def searchRange(self, nums: List[int], target: int) -> List[int]:
        n = len(nums)
        ans = [0, 0]

        if not nums:
            return [-1, -1]

        def dfs(t1):
            left = 0
            right = n
            while left < right:
                mid = (left + right) // 2

                if nums[mid] < t1:
                    left = mid + 1
                else:
                    right = mid

            return left

        ans[0] = dfs(target)
        ans[1] = dfs(target+1)-1

        return ans if ans[0] <= ans[1] else [-1, -1]
```

# **1334**

```python
class Solution:
    def findTheCity(self, n: int, edges: List[List[int]], distanceThreshold: int) -> int:
        arr = [[float('inf') for _ in range(n)] for _ in range(n)]

        for start, end, w in edges:
            arr[start][end] = w
            arr[end][start] = w

        for i in range(n):
            arr[i][i] = 0

        for i in range(n):
            for j in range(n):
                for k in range(n):
                    arr[j][k] = min(arr[j][k], arr[j][i] + arr[i][k])

        ans = [0] * n
        for i in range(n):
            res = -1
            for d in arr[i]:
                if d <= distanceThreshold:
                    res += 1
            ans[i] = res

        return n-1-ans[::-1].index(min(ans))
```

# **2976**

플루이드

```python
class Solution:
    # 26 : alphabet 개수
    def minimumCost(self, source: str, target: str, original: List[str], changed: List[str], cost: List[int]) -> int:
        conversionGraph = self.buildConversionGraph(original, changed, cost)
        self.optimizeConversionPaths(conversionGraph)
        return self.computeTotalConversionCost(source, target, conversionGraph)


    def buildConversionGraph(self, original: List[str], changed:List[str], cost: List[int]) -> List[List[int]]:
        graph = [[float('inf')] * 26 for _ in range(26)]

        for i in range(26):
            graph[i][i] = 0
        for i in range(len(cost)):
            fromChar = ord(original[i]) - ord('a')
            toChar = ord(changed[i]) - ord('a')
            graph[fromChar][toChar] = min(graph[fromChar][toChar], cost[i])     # cost 있는 값들 삽입.
        return graph

    def optimizeConversionPaths(self, graph: List[List[int]]) -> None:
        for k in range(26):
            for i in range(26):
                if graph[i][k] < float('inf'):                                  # cost 에 있는 값들인지 확인
                    for j in range(26):
                        if graph[k][j] < float('inf'):
                            graph[i][j] = min(graph[i][j], graph[i][k] + graph[k][j])       # 플루이드 알고리즘

    def computeTotalConversionCost(self, source: str, target: str, graph: List[List[int]]) -> int:
        totalCost = 0
        for i in range(len(source)):
            sourceChar = ord(source[i]) - ord('a')
            targetChar = ord(target[i]) - ord('a')
            if sourceChar != targetChar:
                if graph[sourceChar][targetChar] == float('inf'):
                    return -1
                totalCost += graph[sourceChar][targetChar]
        return totalCost
```

# **2045**

```python
class Solution:
    def secondMinimum(self, n: int, edges: List[List[int]], time: int, change: int) -> int:
        D = [[] for _ in range(n+1)]
        D[1] = [0]
        G, heap = defaultdict(list), [(0, 1)]

        for a, b in edges:
            G[a] += [b]
            G[b] += [a]


        while heap:
            min_dist, idx = heappop(heap)
            if idx == n and len(D[n]) == 2: return max(D[n])

            for neib in G[idx]:
                if (min_dist // change) % 2 == 0:
                    cand = min_dist + time
                else:
                    cand = ceil(min_dist/(2*change)) * (2*change) + time

                if not D[neib] or (len(D[neib]) == 1 and D[neib] != [cand]):
                    D[neib] += [cand]
                    heappush(heap, (cand, neib))
```

```python
class Solution:
    def minSwaps(self, nums: List[int]) -> int:
        ones = nums.count(1)
        nums = nums + nums
        x, onesInWindow = 0, 0
        for i in range(len(nums)):
            if i >= ones and nums[i - ones]: x -= 1
            if nums[i] == 1: x += 1
            onesInWindow = max(x, onesInWindow)
        return ones - onesInWindow

```

# **959**

DFS

```python
class Solution:
    def regionsBySlashes(self, grid: List[str]) -> int:
        # kick : / -> 3*3 배열으로 표현

        n = len(grid)
        ans = 1
        dp = [[0 for _ in range(n*3)] for _ in range(n*3)]
        dx = [-1, 1, 0, 0]
        dy = [0, 0, -1, 1]

        for i in range(n):
            for j in range(len(grid[i])):
                if grid[i][j] == "/":
                    dp[i*3][j*3+2] = -1
                    dp[i*3+1][j*3+1] = -1
                    dp[i*3+2][j*3] = -1
                elif grid[i][j] == "\\":
                    dp[i*3][j*3] = -1
                    dp[i*3+1][j*3+1] = -1
                    dp[i*3+2][j*3+2] = -1

        def dfs(y, x, ans):
            for i in range(4):
                nx = x + dx[i]
                ny = y + dy[i]

                if 0 <= nx < 3*n and 0 <= ny < 3*n and dp[ny][nx] == 0:
                    dp[ny][nx] = ans
                    dfs(ny, nx, ans)

        for i in range(3*n):
            for j in range(3*n):
                if dp[i][j] == 0:
                    dfs(i, j, ans)
                    ans += 1

        return ans-1
```

```python
class KthLargest:
    import heapq

    def __init__(self, k: int, nums: List[int]):
        self.hq = []
        self.a = k

        for num in nums:
            self.add(num)

    def add(self, val: int) -> int:
        heappush(self.hq, val)
        if len(self.hq) > self.a:
            heappop(self.hq)
        return self.hq[0]


# Your KthLargest object will be instantiated and called as such:
# obj = KthLargest(k, nums)
# param_1 = obj.add(val)
```
