# **ZOAC4**

구현

```python
import sys
import math

H, W, N, M = list(map(int, sys.stdin.readline().split(' ')))

ans = math.ceil(H/(N+1)) * math.ceil(W/(M+1))

print(ans)
```

# **삼각형과 세 변**

구현

```python
import sys
input = sys.stdin.readline

while True:
    a, b, c = map(int, input().split())
    if a == b == c == 0:
        break

    if max(a, b, c) * 2 >= a + b + c:
        print("Invalid")
    else:
        if a == b == c:
            print("Equilateral")
        elif a == b or b == c or a == c:
            print("Isosceles")
        else:
            print("Scalene")
```

# **단어 공부**

dict

```python
import sys

input = sys.stdin.readline().strip()
input = input.upper()

dic = {}
for i in range(len(input)):
    dic[input[i]] = dic.get(input[i], 0) + 1

dic = sorted(dic.items(), key=lambda x: -x[1])
if len(dic) > 1 and dic[0][1] == dic[1][1]:
    print("?")
else:
    print(dic[0][0])
```
